**Q 1 - Write a program that take a user input of three angles and will find out whether it can form a triangle or not.**

```python
# take three angles from user
first = int(input("Enter the 1st angle of a triangle: "))
second = int(input("Enter the 2nd angle of a triangle: "))
third = int(input("Enter the 3rd angle of a triangle: "))

# check triangle condition
if (first + second + third == 180) and first > 0 and second > 0 and third > 0:
  print("Forms a triangle")
else:
  print("Does not forms a triangle")
```

**Q 2 - Write a program that will take user input of cost price and selling price and determines whether its a loss or a profit.**

```python
# take cost price and selling price from user
cp = float(input("Enter the cost price: "))
sp = float(input("Enter the selling price: "))

# check profit or loss
if sp > cp:
  profit = sp - cp
  print("Profit =",profit)

elif sp < cp:
  loss = cp - sp
  print("Loss =",loss)

else:
  print("No Profit, No Loss")
```

**Q 3 - Write a menu-driven program**

1.   cm to ft
2.   km to miles
1.   USD to INR
2.   exit

```python
# show menu and take choice from user
menu = input("""
Hey, select an option:

1. cm to ft
2. km to miles
3. USD to INR
4. Exit

Enter your choice: """)

# option 1: cm to ft
if menu == "1":
  cm = int(input("Enter the value in centimeters: "))
  ft = cm / 30.48
  print("Value in feet =",round(ft,2))

# option 2: km to miles
elif menu == "2":
  km = int(input("Enter the value in kilometers: "))
  miles = km * 0.621371
  print("Distance in miles =",round(miles,2))

# option 3: USD to INR
elif menu == "3":
  usd = int(input("Enter the amount in USD: "))
  inr = usd * 83
  print("Amount in INR =",round(inr,2))

# option 4: Exit
elif menu == "4":
  print("Exit")

# invalid option
else:
  print("Invalid choice. Please select a vaild option.")
```

**Q 4 - Display Fibonacci series up to 10 terms.**

```python
# initialize the first two numbers
num1 = 0
num2 = 1

# loop to print 10 terms
for i in range(10):
  # print current number
  print(num1, end=" ")

  # calculate next number
  next_num = num1 + num2

  # update values
  num1 = num2
  num2 = next_num
```

**Q 5 - Find the factorial of a given number.**

```python
# take number from user
num = int(input("Enter a number: "))

# initialixe factorial variable
factorial = 1

# loop from 1 to num
for i in range(1, num+1):
  factorial = factorial * i

# print the result
print("Factorial =",factorial)
```

**Q 6 - Reverse a given integer number.**

```python
# take the number from the user
num = int(input("Enter a number: "))

reverse = 0

# loop runs until the number becomes 0
while num > 0:

    # extract the last digit
    last_digit = num % 10
    reverse = reverse * 10 + last_digit
    num = num // 10

# display the reversed number
print("Reversed number =", reverse)
```

**Q 7 - Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum futher more. Print the final result. And don't use for loop to solve this problem.**

```python
# take input from user
N = int(input("Enter the number: "))

total = 0
i = 1

while i <= N:

    # skip numbers divisible by 5
    if i % 5 == 0:
        i += 1
        continue

    total += i

    # stop if total becomes greater than 300
    if total > 300:
        total -= i
        break

    i += 1

print("Final Sum =", total)
```

**Q 8 - Write a program that keeps on accepting a number from the user until the user enters Zero. Display the sum and average of all the numebers.**

```python
total = 0
count = 0

while True:
    num = int(input("Enter a number (0 to stop): "))

    # stop the loop if user enters 0
    if num == 0:
        break

    total += num
    count += 1

# calculate average
if count != 0:
    average = total / count
else:
    average = 0

# display results
print("Sum =", total)
print("Average =", average)
```

**Q 9 - Write a program which will find all such numbers which are  divisible by 7 but are not a multiple of 5, between 2000 and 3000 (both included). The numbers obtained should be printed in a comma-sepatated sequence on a single line.**

```python
numbers = []

for i in range(2000, 3001):
    if i % 7 == 0 and i % 5 != 0:
        numbers.append(str(i))

print(",".join(numbers))
```
