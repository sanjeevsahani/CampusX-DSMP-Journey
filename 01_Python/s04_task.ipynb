**Q 1 - Combine two lists index-wise (columns wise).**

    Write a program to add two lists index-wise. Create a new list that contains the 0th index item from both the list,
    then the 1st index item, and so on till element. any leftover items will get added at the end of the new list.

    Given List:
               List1 = ["M", "na", "i", "kh"]
               List2 = ["y", "me", "s", "an"]

    Output: [["M", "y"], ["na", "me"], ["i", "s"], ["kh", "an"]]

```python
# Using zip()

list1 = ["M", "na", "i", "kh"]
list2 = ["y", "me", "s", "an"]

L = []

for i,j in zip(list1,list2):
  L.append([i,j])

print(L)
```

```python
# Using List Comprehension

list1 = ["M", "na", "i", "kh"]
list2 = ["y", "me", "s", "an"]

L = [[i,j] for i,j in zip(list1,list2)]
print(L)
```

**Q 4 - Running Sum on list Write a program to print a list after performing running sum on it.**

    Input: list1 = [1,2,3,4,5,6]
    
    Output: [1,3,6,10,15,21]

```python
list1 = [1,2,3,4,5,6]

# Empty list to store result
result = []
total = 0

for i in list1:
  total += i
  result.append(total)

# Print final result
print(result)
```

**Q 5 - You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.**

    i.e. Say given list is [2,4,6,10,1] resultant list will be [22,20,10,23].

    For 1st element 2 ->> these are greater (4+6+10) values and 2 itself so on adding becomes 22.

    For 2nd element 4 ->> greater elements are (6, 10) and 4 itself, so on adding 20 like wise for all other elememts.
    
    [2,4,6,10,1]-->[22,20,16,10,23]

```python
# Input list
list1 = [2, 4, 6, 10, 1]

# List to store result
result = []

# Traverse each element in the list
for i in list1:
    total = 0   # Reset total for each element

    # Compare with all elements in the list
    for j in list1:
        if j >= i:        # Check for greater or equal elements
            total += j    # Add to total

    # Append result for current element
    result.append(total)

# Print final result
print(result)
```

**Q 6 - Find list of common unique items from two list. and show in increasing order.**

    Input: num1 = [23,45,67,78,89,34]
           num2 = [34,89,55,56,39,67]

    Output: [34, 67, 89]

```python
# Input lists
num1 = [23, 45, 67, 78, 89, 34]
num2 = [34, 89, 55, 56, 39, 67]

# List to store common unique elements
common = []

# Traverse first list
for i in num1:
    # Check if element is in second list and not already added
    if i in num2 and i not in common:
        common.append(i)

# Sort the result in ascending order
common.sort()

# Print final result
print(common)
```

**Q 7 - Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.**

    Input: ['1ac21', '23fg', '456', '098d','1','kls']

    Output: ['456', '23fg', '1ac21', '1', 'kls', '098d']

```python
L = ['1ac21', '23fg', '456', '098d', '1', 'kls']

# List to store product values
prod_value = []

# Calculate product of digits for each string
for item in L:
    product = 1

    for char in item:
        if char.isdigit():
            product = product * int(char)

    # Append product (corrected)
    prod_value.append(product)

# Zip product values with original list, sort in descending order
result = [i[1] for i in sorted(list(zip(prod_value, L)), reverse=True)]

# Print final result
print(result)
```

**Q 8 - Split String of list on K character.**

    Input: ['CampusX is a channel', 'for data-science', 'aspirants.']
    utput: ['CampusX', 'is', 'a', 'channel', 'for', 'data-science', 'aspirants.']

```python
# Input list
L = ['CampusX is a channel', 'for data-science', 'aspirants.']

# Empty list to store result
result = []

# Traverse each string in list
for item in L:
    words = item.split()
    result.extend(words)

# Print final result
print(result)
```

**Q 10 - Add Space between Potential Words.**

    Input: ['campusxIs', 'bestFor', 'dataScientist']

    Output: ['campusx Is', 'best For', 'data Scientist']

```python
# Input list
L = ['campusxIs', 'bestFor', 'dataScientist']

result = []

# Traverse each string
for word in L:
    new_word = ""

    for ch in word:
        # Check if character is uppercase
        if ch.isupper():
            new_word += " " + ch
        else:
            new_word += ch

    result.append(new_word)

print(result)
```

**Q 11 - Write a program that can perform union operation on 2 lists**.

    Input: [1,2,3,4,5,1]
           [2,3,5,7,8]

    Output: [1,2,3,4,5,7,8]

```python
L1 = [1, 2, 3, 4, 5, 1]
L2 = [2, 3, 5, 7, 8]

result = []

# Add elements from first list
for i in L1:
    if i not in result:
        result.append(i)

# Add elements from second list
for i in L2:
    if i not in result:
        result.append(i)

# Sort the result
result.sort()

print(result)
```

**Q 12 - Write a program that can find the max number of each row of a matrix.**

    Input: [[1,2,3],[4,5,6],[7,8,9]]
    Output: [3,6,9]

```python
# Input matrix
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

result = []

# Traverse each row
for row in matrix:
    result.append(max(row))

# Print result
print(result)
```

**Q 14 - Write a list comprehension that can transpose a given matrix.**

    matrix = [[1,2,3],
              [4,5,6],
              [7,8,9]]
              
    Outpurt:  [1, 4, 7]
              [2, 5, 8]
              [3, 6, 9]

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

# Transpose using nested list comprehension
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]

print(transpose)
```

**Q 15 - Write a list comprehension that can flatten a nested list.**

    Input: matrix = [[1,2,3],
                     [4,5,6],
                     [7,8,9]]

    Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

# Flatten the matrix
flat_list = [item for row in matrix for item in row]

print(flat_list)
```
