# **Tuples**

**A tuple is an ordered and immutable collection of elements. It is similar to a list, but the main difference is that tuple elements cannot be changed once they are created, whereas list elements can be modified.**

```python
T = (10, 20, 30, 20)

print(T[1])
```

**Characteristics of Tuples**

**1. Ordered**

- Elements are stored in a fixed sequence
- Each element has an index (position)
- ndex starts from 0

**2. Immutable (Unchangeable)**
- Once created, elements cannot be modified, added, or removed

```python
T = (1, 2, 3)
T[0] = 10   # Error
```

**3. Allows Duplicates**
- Same value can appear multiple times

```python
T = (1, 2, 2, 3)
```

**4. Heterogeneous**
- Can store different data types

```python
T = (1, "Hello", 3.5, True)
```

**Why Use Tuples - Tuples are used for storing fixed and unchangeable data efficiently, providing better performance and data safety.**

# **Topic - Difference between Lists and Tuples**

| Feature              | List                                | Tuple                         |
| -------------------- | ----------------------------------- | ----------------------------- |
| **Syntax**           | `[ ]`                               | `( )`                         |
| **Mutability**       | Mutable (can be changed)            | Immutable (cannot be changed) |
| **Speed**            | Slower                              | Faster                        |
| **Memory Usage**     | More memory                         | Less memory                   |
| **Built-in Methods** | Many (`append()`, `remove()`, etc.) | Few (`count()`, `index()`)    |
| **Error Prone**      | More (can change accidentally)      | Less (data is fixed)          |
| **Usability**        | Dynamic data                        | Fixed data                    |


**Use List when:**
- Data changes frequently
- Need to add/remove elements

**Use Tuple when:**
- Data should remain constant
- Need better performance
- Using as dictionary key

**Similarities**
- Both are ordered collections
- Allow duplicates
- Support indexing and slicing

```python
# List --> Mutability

a = [1, 2, 3]
b = a

a.append(4)
print(a)
print(b)
```

```python
# Tuple  --> Mutability

a = (1, 2, 3)
b = a

a = a + (4,)
print(a)
print(b)
```

```python
# speed

import time

L = list(range(1000000))
T = tuple(range(1000000))

start = time.time()
for i in L:
  i * 5
print("List time :",time.time()-start)

start = time.time()
for i in T:
  i * 5
print("Tuple time :",time.time()-start)
```

```python
# memory

import sys

L = list(range(1000))
T = tuple(range(1000))

print("List size: ",sys.getsizeof(L))
print("Tuple size: ",sys.getsizeof(T))
```

# **Topic - Accessing Elements in Tuples**

**In Python, a tuple elements are accessed using indexing and slicing, just like lists.**

**Positive Indexing**
- Each element has a position (index)
- Index starts from 0

```python
t = (10, 20, 30, 40, 50)

print(t[0])
print(t[1])
print(t[4])
```

**Negative Indexing**
- Used toaccess elements from the end

```python
t = (1, 2, 3, 4, 5)

print(t[-1])
print(t[-3])
print(t[-5])
```

**Slicing**
- Used to access a range of elements

```python
# positive

t = (1, 2, 3, 4, 5)

print(t[0:5])
print(t[1:3])
print(t[:])
print(t[:4])
print(t[2:])
print(t[::2])
```

```python
# Negative

t = (1, 2, 3, 4, 5)

print(t[-5:-1])
print(t[:-1])
print(t[-5:])
print(t[::-2])
```

```python
# inverse

t = (1, 2, 3, 4, 5)

print(t[::-1])
```

```python
# Nested tuple

t = (1, 2, (3, 4, 5))

print(t[1])
print(t[2])
print(t[2][0])
print(t[2][1])
print(t[2][2])
```

**Important Points**
- Use [] to access elements
- Index must be integer
- Out-of-range index → IndexError
- Tuples are immutable, but accessing is allowed

# **Topic - Set Operations vs Update Methods**

**In Python, every major set operation has two forms:**

- Normal method → returns a new set
- Update method → modifies the original set

**1. Union vs Update**

| Method     | Description                         |
| ---------- | ----------------------------------- |
| `union()`  | Returns a new set with all elements |
| `update()` | Adds elements to original set       |


```python
A = {1, 2}
B = {2, 3}

print(A.union(B))
```

```python
A = {1, 2}
B = {2, 3}

A.update(B)
print(A)
print(B)
```

**2. Intersection vs Intersection Update**

| Method                  | Description                                |
| ----------------------- | ------------------------------------------ |
| `intersection()`        | Returns common elements                    |
| `intersection_update()` | Keeps only common elements in original set |


```python
A = {1, 2, 3}
B = {2, 3, 4}

print(A.intersection(B))
```

```python
A = {1, 2, 3}
B = {2, 3, 4}

A.intersection_update(B)
print(A)
print(B)
```

**3. Difference vs Difference Update**

| Method                | Description                               |
| --------------------- | ----------------------------------------- |
| `difference()`        | Returns elements in A not in B            |
| `difference_update()` | Removes common elements from original set |


```python
A = {1, 2, 3}
B = {2, 3}

print(A.difference(B))
```

```python
A = {1, 2, 3}
B = {2, 3}

A.difference_update(B)
print(A)
print(B)
```

**4. Symmetric Difference vs Symmetric Difference Update**

| Method                          | Description                                   |
| ------------------------------- | --------------------------------------------- |
| `symmetric_difference()`        | Elements in either A or B but not both        |
| `symmetric_difference_update()` | Updates original set with non-common elements |


```python
A = {1, 2, 3}
B = {3, 4}

print(A.symmetric_difference(B))
```

```python
A = {1, 2, 3}
B = {3, 4}

A.symmetric_difference_update(B)
print(A)
print(B)
```

**Key Concept (Very Important)**

**Normal Methods**

- Return new set
- Original set remains unchanged

**Update Methods**

- Modify the original set
- Return None

# **Dictionary**

**A dictionary in Python is a collection of key-value pairs, used to store data in a structured form, similar to a map. Unlike other data types that store a single value, a dictionary stores data as pairs of keys and corresponding values.**

```python
# Syntax

D = {key1: value1, key2: value2}
```

```python
D = {"name":"Sanjeev", "age":22}

print(D)
```

```python
print(D["name"])
```

**Characteristics of Dictionary**

**1. Key-Value Pair Structure**
- Data is stored in the form of key → value
- Each key is associated with a value

```python
D = {"name":"Sanjeev", "age":22}
```

**2. Keys Must Be Unique**
- Duplicate keys are not allowed

```python
D = {"a": 1, "a": 2}
print(D)
```

**3. Mutable**
- Dictionary is changeable
- You can add, update, or delete elements

```python
D = {"a": 1}
D["a"] = 100

print(D)
```

**4. Unordered (Conceptually)**
- Elements are not accessed using index
- Access is done using keys

**5. Heterogeneous**
- Can store different data types

```python
D = {1: "one", "two": 2, 3: [1, 2]}

print(D)
```

**6. Keys Must Be Immutable**
- Allowed: int, string, tuple
- Not allowed: list, set

```python
D = {(1, 2): "value"}   # ✔
# D = {[1,2]: "value"}  ❌
```

**7. Values Can Be Any Type**

# **Topic - Adding / Updating Elements in Dictionary**

**In Python, elements in a dictionary can be added or updated by assigning values to keys or using the update() method.**

**Adding New Element**
- Use a new key

```python
D = {"name":"Sanjeev", "age":22}

D["city"] = "Ballia"
print(D)
```

**Updating Existing Element**
- Use an existing key

```python
D = {"name":"Sanjeev", "age":22, "city":"Ballia"}

D["city"] = "Delhi"
print(D)
```

**Using update() Method**
- Add or update multiple elements

```python
D = {"a":1}

D.update({"b":2, "c":3})
print(D)
```

```python
D = {"a": 1}

D["a"] = 100   # updates, not duplicate
print(D)
```

**Important Points**
- Dictionaries are mutable
- Adding and updating use same syntax
- update() can add multiple items at once

# **Topic - Deleting Elements in Dictionary**

**In Python, dictionary elements can be deleted using del, pop(), popitem(), and clear() depending on the requirement.**

**Using del Keyword**
- Deletes a specific key-value pair

```python
D = {"a": 1, "b": 2}

del D["a"]
print(D)
```

**Using pop() Method**
- Removes element and returns its value

```python
D = {"a": 1, "b": 2}

value = D.pop("a")
print(value)
print(D)
```

**Using popitem() Method**
- Removes the last inserted key-value pair

```python
D = {"a": 1, "b": 2}

D.popitem()
print(D)
```

**Using clear() Method**
- Removes all elements

```python
D = {"a": 1, "b": 2}

D.clear()
print(D)
```

**Deleting Entire Dictionary**

```python
D = {"a": 1}

del D
```

**Important Points**
- del → removes specific item
- pop() → removes + returns value
- popitem() → removes last item
- clear() → empties dictionary

# **Topic - Dictionary Operations**

**In Python, dictionaries support various operations to manipulate and work with key-value pairs.**

**Membership Operations**
- Checks if a key exists

```python
D = {"a": 1, "b": 2}

print("a" in D)
print("b" in D)

print("c" in D)
```

**Works on keys only, not values**

```python
D = {"a": 1, "b": 2}

print("a" not in D)
print("b" not in D)

print("c" not in D)
```

**Iteration (Looping)**

```python
# Loop through keys

D = {"a": 1, "b": 2}

for key in D:
  print(key)
```

```python
# Loop through values

D = {"a": 1, "b": 2}

for value in D.values():
  print(value)
```

```python
# Loop through key-value pairs

D = {"a": 1, "b": 2}

for key, value in D.items():
  print(key, value)
```

# **Topic - Dictionary Functions**

**In Python, dictionary functions like len(), max(), min(), and sorted() operate mainly on keys and are used for efficient data handling.**

**len() - Returns number of key-value pairs**

```python
D = {"a": 1, "b": 2}

print(len(D))
```

**max() - Returns the maximum key**

```python
D = {"a": 1, "b": 2, "c": 3}

print(max(D))
```

**min() - Returns the minimum key**

```python
D = {"a": 1, "b": 2, "c": 3}

print(min(D))
```

**sum() - Returns sum of keys (only if numeric)**

```python
D = {1: "a", 2: "b", 3: "c"}

print(sum(D))
```

**sorted() - Returns a sorted list of keys**

```python
# accending order

D = {"b": 2, "a": 1}

print(sorted(D))
```

```python
# Decending order

D = {"b": 2, "a": 1}

print(sorted(D,reverse=True))   # ['a', 'b']
```

**all() - Returns True if all keys are True**

```python
D = {1: "a", 2: "b"}

print(all(D))
```

**any() - Returns True if any key is True**

```python
D = {0: "a", 2: "b"}

print(any(D))
```

**dict() - Creates a dictionary**

```python
D = dict(a=1, b=2)
print(D)
```

**keys() - Returns all the keys of the dictionary.**

```python
D = {"a": 1, "b": 2}

print(D.keys())
```

**values() - Returns all the values of the dictionary.**

```python
D = {"a": 1, "b": 2}

print(D.values())
```

**items() - Returns all key-value pairs as tuples.**

```python
D = {"a": 1, "b": 2}

print(D.items())
```

# **Topic - Dictionary Comprehension**

**Dictionary comprehension is a compact syntax used to create a dictionary by generating key-value pairs from an iterable.**

```python
# Syntax

{key_expression: value_expression for item in iterable if condition}
```

```python
D = {x: x*x for x in range(5)}
print(D)
```

```python
D = {x: x*x for x in range(10) if x % 2 == 0}
print(D)
```

```python
L = ["a", "b", "c"]

D = {x: len(x) for x in L}
print(D)
```

```python
# Print 1st 10 numbers and their squares

D = {i: i**2 for i in range(1,11)}
print(D)
```
