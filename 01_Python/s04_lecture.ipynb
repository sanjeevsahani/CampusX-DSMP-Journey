# **Topic - Introduction to Lists**

**What is Lists - A list is an ordered, mutable collection of elements that can store multiple items under one name. In languages like Python, lists are implemented as dynamic arrays, allowing elements to be added or removed at runtime.**

```python
L = [20, "Jessa", 35.75, [30, 60, 90]]
```

**Why Lists are required in programing?**

Lists are required because they:

- Reduce code complexity
- Handle multiple values efficiently
- Support dynamic and real-world data
- Enable looping and data processing
- Form the basis of advanced programming concepts

**Array vs Lists**

| Feature     | Array          | List                  |
| ----------- | -------------- | --------------------- |
| Size        | Fixed          | Dynamic               |
| Data Type   | Same type only | Mixed types allowed   |
| Memory      | Contiguous     | Dynamic allocation    |
| Flexibility | Low            | High                  |
| Speed       | Faster         | Slightly slower       |
| Operations  | Limited        | Many built-in methods |


**How Lists are stored in memory**
- In Python, a list is stored as a dynamic array of references, where each element holds the memory address of an object rather than the actual value.

```python
L = [1,2,3]

print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))

print(id(1))
print(id(2))
print(id(3))
```

**Characterstics of a List**
- Ordered
- Changeble / Mutable
- Hetrogeneous
- Can have duplicates
- are dynamic
- can be nested
- items can be accessed
- can contain any kind of objects in python

```python
# Ordered

L = [10, 20, 30]
print(L[0])  # 10
```

```python
# Mutable (Changeable)

L = [10, 20, 30]
L[1] = 99
print(L)   # [10, 99, 30]
```

```python
# Allows Duplicate Values

L = [10, 20, 10, 30]
```

```python
# Heterogeneous (Different Data Types Allowed)

L = [10, "Hello", 3.5, True]
```

```python
# Dynamic Size

L = [1, 2]
L.append(3)   # Add element
L.remove(1)   # Remove element
print(L)
```

```python
# Supports Indexing and Slicing

L = [10, 20, 30, 40]
print(L[1:3])   # [20, 30]
```

```python
# Can Store Nested Lists

L = [1, 2, [3, 4]]
```

```python
# Iterable

for item in [1, 2, 3]:
    print(item)
```

# **Topic - Accessing Elements in Lists**


**Positive Indexing - Each element has a position (index). Index starts from 0**

1D List

- Elements are accessed using a single index: L[i]

2D List

- Elements are accessed using two indices: L[row][column]

3D List

- Elements are accessed using three indices: L[block][row][column]

```python
# Accessing Elements in 1D List

L = [10, 20, 30, 40]

print(L[0])   # 10
print(L[2])   # 30
```

```python
# Accessing Elements in 2D List

L = [[1, 2, 3], [4, 5, 6]]

print(L[0])        # [1, 2, 3]
print(L[1][2])     # 6
```

```python
# Accessing Elements in 3D List

L = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

print(L[1])
print(L[1][0])
print(L[1][0][1])   # 6
```

**Negative Indexing - Used to access elements from the end.**

```python
# Accessing Elements in 1D List

L = [10, 20, 30, 40]

print(L[-1])   # 40
print(L[-2])   # 30
```

```python
# Accessing Elements in 2D List

L = [[1, 2, 3], [4, 5, 6]]

print(L[-1])        # [4, 5, 6]
print(L[-1][-1])     # 6
```

```python
# Accessing Elements in 3D List

L = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

print(L[-1])
print(L[-1][-2])
print(L[-1][-2][-1])   # 6
```

**Slicing - Slicing is used to extract a portion of a list using the syntax L[start:end:step], where the start index is included and the end index is excluded.**

**Basic Syntax - L[start : end : step]**
- start → index where slicing begins (included)
- end → index where slicing stops (excluded)
- step → interval between elements

```python
# Positive Slicing

L = [10, 20, 30, 40, 50]

print(L[1:4])   # [20, 30, 40]
print(L[0:3])   # [10, 20, 30]
print(L[:3])    # [10, 20, 30]   (start = 0)
print(L[2:])    # [30, 40, 50]   (end = last)
print(L[:])     # [10, 20, 30, 40, 50] (full list)
```

```python
L = [10, 20, 30, 40, 50]

print(L[::2])   # [10, 30, 50]
print(L[1::2])  # [20, 40]
```

```python
# Negitive Slicing

L = [10, 20, 30, 40, 50]

print(L[-4:-1])   # [20, 30, 40]
print(L[-5:-2])   # [10, 20, 30]
print(L[-5:-1])     # [10, 20, 30, 40]
```

```python
L = [10, 20, 30, 40, 50]

print(L[::-1])     # [50, 40, 30, 20, 10]
print(L[::-2])     # [50, 30, 10]
print(L[-1::-2])   # [50, 30, 10]
```

```python
# Reverse a List

L = [10, 20, 30, 40]

print(L[::-1])   # [40, 30, 20, 10]
```

# **Topic - Adding Elements in Lists**


**Elements can be added to a list using methods like append(), insert(), and extend(), which allow adding elements at the end, at a specific position, or multiple elements respectively.**

**append()**
- Adds an element at the end of the list.

```python
L = [10, 20, 30]

L.append(40)
print(L)   # [10, 20, 30, 40]
```

```python
L = [1,2,3,4,5]
L.append(True)
print(L)
```

```python
L = [1,2,3,4,5]
L.append([6,7,8])
print(L)
```

**insert()**
- Adds an element at a specific position (index)
- Syntax - L.insert(index, element)

```python
L = [10, 20, 30]

L.insert(1, 15)
print(L)   # [10, 15, 20, 30]
```

```python
L = [1,2,3,4,5]
L.insert(5,"Ballia")
print(L)
```

**extend()**
- Adds multiple elements (from another list)

```python
L = [10, 20]

L.extend([30, 40])
print(L)   # [10, 20, 30, 40]
```

```python
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)
```

```python
L = [1,2,3,4,5]
L.extend("ballia")
print(L)
```

# **Topic - Updating Elements in Lists**

**Elements in a list can be updated using indexing or slicing because lists are mutable in Python.**

```python
# Updating a Single Element (Using Index)

L = [10, 20, 30, 40]

L[1] = 99
print(L)   # [10, 99, 30, 40]
```

```python
# Updating Multiple Elements (Using Slicing)

L = [10, 20, 30, 40]

L[1:3] = [100, 200]
print(L)   # [10, 100, 200, 40]
```

```python
# Replacing with Different Number of Elements

L = [10, 20, 30, 40]

L[1:3] = [100]
print(L)   # [10, 100, 40]
```

```python
# Updating Nested List Elements

L = [[1, 2], [3, 4]]

L[1][0] = 99
print(L)   # [[1, 2], [99, 4]]
```

```python
# Using Loop to Update Elements

L = [1, 2, 3]

for i in range(len(L)):
    L[i] = L[i] * 2

print(L)   # [2, 4, 6]
```

# **Topic - Removing / Deleting Elements from Lists**


**Elements can be removed from a list using remove(), pop(), del, and clear(), depending on whether removal is based on value, index, or entire list.**

| Method   | Removes by | Returns value | Notes                 |
| -------- | ---------- | ------------- | --------------------- |
| remove() | value      | No            | First occurrence only |
| pop()    | index      | Yes           | Default last element  |
| del      | index      | No            | Can delete slice      |
| clear()  | all        | No            | Empties list          |


**remove()**
- Removes the first occurrence of a value

```python
L = [1, 2, 3, 4, 5]

L.remove(5)
print(L)
```

```python
L = [10, 20, 30, 20]

L.remove(20)
print(L)   # [10, 30, 20]
```

**pop()**
- Removes element using index
- Also returns the removed value

```python
L = [10, 20, 30]

L.pop(1)
print(L)   # [10, 30]
```

```python
# Default remove the last element
L = [10, 20, 30]

L.pop()
print(L)
```

**del**
- Deletes element by index or slice

```python
L = [10, 20, 30, 40]

del L[1]
print(L)   # [10, 30, 40]
```

```python
L = [1, 2, 3, 4, 5]

del L[-1]
print(L)
```

```python
L = [10, 20, 30, 40]

del L[1:3]
print(L)
```

```python
L = [1, 2, 3, 4, 5]

del L[-5:-2]
print(L)
```

```python
# complete delete

L = [1, 2, 3, 4, 5]

del L
print(L)
```

**clear()**
- Removes all elements

```python
L = [1, 2, 3, 4, 5]

L.clear()
print(L)
```

```python
L = [10, 20, 30]

L.clear()
print(L)   # []
```
