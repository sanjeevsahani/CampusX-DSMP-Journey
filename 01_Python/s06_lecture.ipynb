# **Functions**

**A function is a named block of code that is defined once and can be executed multiple times by calling it. It can take inputs (parameters), process them, and return an output.**

```python
# Syntax

def function_name(parameters):
    # code block
    return value
```

```python
# Function to check whether a number is even or odd
def is_even(num):
    """
    This function checks whether a given number is even or odd.

    Input  : Integer number
    Output : "Even" or "Odd"
    """

    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"


# Calling function for numbers from 1 to 10
for i in range(1, 11):
    result = is_even(i)
    print(f"{i} -> {result}")
```

```python
print(is_even.__doc__)
```

**Types of Functions**

    1 - Built-in Functions
        Already defined in Python
        Example: len(), max(), print()
        
    2 - User-defined Functions
        Created by user using def

**Parameters - Parameters are variables defined in the function definition.**

**Arguments - Arguments are the actual values passed to the function when it is called.**

```python
def add(a, b):   # a, b → parameters
    return a + b

add(2, 3)        # 2, 3 → arguments
```

**Difference Between**

| Feature    | Parameters                       | Arguments                          |
| ---------- | -------------------------------- | ---------------------------------- |
| Definition | Variables in function definition | Values passed during function call |
| Where used | Inside function declaration      | While calling function             |
| Example    | `def f(x)` → x is parameter      | `f(10)` → 10 is argument           |
| Role       | Placeholder                      | Actual data                        |


# **Types of Arguments**

**1. Positional Arguments**

- Arguments passed in the same order as parameters.

```python
def power(a, b):
    return a ** b

print(power(2, 3))   # a=2, b=3
```

**2. Keyword Arguments**
- Arguments passed using parameter names.

```python
def power(a, b):
    return a ** b

print(power(b=3, a=2))   # Order does not matter
```

**3. Default Arguments**
- Parameters having default values if no argument is passed.

```python
def greet(name="Guest"):
    print("Hello", name)

greet()         # Hello Guest
greet("Sanjeev")   # Hello sanjeev
```

4. Variable Length Arguments(*args)

- Used when number of arguments is not fixed.

```python
def total(*nums):
    return sum(nums)

print(total(1, 2, 3, 4))   # 10
```

**Stored as tuple**

```python
def multiply(*args):
  product = 1

  for i in args:
    product = product * i

  print(args)
  return product

multiply(1,2,3,4,5)
```

5. Keyword Variable Length Arguments (**kwargs)
- Accepts multiple keyword arguments.

```python
def display(**data):
    print(data)

display(name="Aman", age=20)
```

**Stored as dictionary**

```python
def display(**kwargs):

  for (key,value) in kwargs.items():
    print(key, "->", value)

display(india="delhi", srilanka="colombo", nepal="kathmandu", pakistan="islamabad")
```

Points to remember while using *args and **kwargs
- order of the arguments matter (positional → default → *args → **kwargs)
- The words "args" and "kwargs" are only a convention, you can use any name of your choice.

# **How Functions are Executed in Memory**

**Functions in Python are executed using a call stack where each function call creates a stack frame that is removed after execution.**

```python
# Function to check whether a number is even or odd
def is_even(num):

    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(is_even(7))
```

**Important Points**
- Each function call → new stack frame
- Local variables exist only during execution
- After return → memory is freed
- Nested calls → multiple stack frames

**Without return statement**

**If a function does not have a return statement, Python automatically returns None after execution while following the same call stack process.**

```python
def is_even(num):
  if num % 2 == 0:
    print("Even")
  else:
    print("Odd")

print(is_even(7))
```

**Important Points**
- return is optional
- Without return → function returns None
- Used when function only prints or performs action

**Difference (With vs Without Return)**
| Feature | With `return`   | Without `return`    |
| ------- | --------------- | ------------------- |
| Output  | Value returned  | `None` returned     |
| Use     | Data processing | Display/side-effect |
| Example | `return a+b`    | `print(a+b)`        |


# **Variable Scope**

**Variable scope determines where a variable can be used or accessed in a program.**

**Types of Scope**

**1) Local Scope - Variable defined inside a function**

```python
def func():
    x = 10   # local variable
    print(x)

func()
# print(x) Error
```

**Accessible only inside the function**

**2) Global Scope - Variable defined outside all functions**

```python
x = 100   # global variable

def func():
    print(x)

func()
print(x)
```

**Accessible everywhere**

**3) Enclosing Scope (Nonlocal) - Variable in nested function**

```python
def outer():
    x = 10

    def inner():
        print(x)

    inner()

outer()
```

**Inner function can access outer function variables**

**4) Built-in Scope - Predefined names in Python**

```python
print(len("Hello"))   # len is built-in
```

**LEGB Rule - Python searches variables in this order:**

    Local → Enclosing → Global → Built-in

**Important Points**
- Local variables → inside function
- Global variables → outside function
- LEGB rule → search order

# **Nested Functions**

**A nested function is a function that is defined within another function and is accessible only inside that outer function.**

```python
def outer():
  print("This is outer function")

  def inner():
    print("This is inner function")

  inner()

outer()
```

- Inner function is not accessible outside.
- Must be called inside outer function.

```python
def outer():
  x = 10

  def inner():
    print(x)

  inner()

outer()
```

- Inner function can access outer function variables

```python
def outer():
  x = 10

  def inner():
    nonlocal x
    x = 20

  inner()
  print(x)

outer()
```

```python
def outer():
  def inner():
    print("Hello")
  return inner

f = outer()
f()
```

```python
def g(x):
  def h():
    x = "abc"

  x = x + 1
  print("in g(x): x =",x)
  h()
  return x

x = 3
z = g(x)
```

```python
def g(x):
  def h(x):
    x = x + 1
    print("in h(x): x = ",x)
  x = x + 1
  print("in g(x): x = ",x)
  h(x)
  return x

x = 3
z = g(x)
print("in main program scope: x = ",x)
print("in main program scope: z = ",z)
```

# **Lambda Function**

**In Python, a lambda function is a small anonymous (nameless) function used for short operations.**

```python
# Syntax

lambda arguments: expression
```

```python
add = lambda a, b: a + b

print(add(2, 3))
```

```python
# x --> x^2

power = lambda x: x**2
print(power(2))
```

```python
print((lambda x: x*x)(5))
```

```python
multiply = lambda a, b, c: a * b * c

print(multiply(2, 3, 4))
```

**Difference Between Lambda vs Normal Function**

- No name
- lambda has no return value(infact, returns a function)
- lambda is written in 1 line
- not reusable

**Then why use lambda function?**

**They are used eith HOF**

```python
# check if a string has 'a'

a = lambda s: "a" in s
a("hello")
```

```python
# odd or even

a = lambda x: "even" if x%2 == 0 else "odd"

print(a(6))
print(a(7))
```

**Higher Order Functions (HOF)**

**A Higher Order Function is a function that takes another function as input or returns a function as output.**

```python
def square(x):
  return x*x

def apply(func, value):
  return func(value)

print(apply(square, 5))
```

```python
def outer():
  def inner():
    return "Hello"
  return inner

f = outer()
print(f())
```

```python
def square(x):
  return x**2

# HOF
def transform(f,L):
  output = []
  for i in L:
    output.append(f(i))

  print(output)

L = [1,2,3,4,5]

transform(square,L)
```

```python
L = [1,2,3,4,5]

transform(lambda x: x**2,L)
```

# **Map, Filter and Reduce**

**map() applies a given function to each item of an iterable and returns a map object (iterator).**

```python
# Syntax

map(function, iterable)
```

```python
def square(x):
    return x * x

nums = [1, 2, 3, 4]

result = map(square, nums)

print(list(result))
```

```python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x*x, nums))

print(result)
```

```python
a = [1, 2, 3]
b = [4, 5, 6]

result = list(map(lambda x, y: x + y, a, b))

print(result)
```

```python
# square the items of a list

list(map(lambda x: x**2,[1,2,3,4,5]))
```

```python
# odd / even labelling of list items

L = [1,2,3,4,5]
list(map(lambda x: "even" if x%2 == 0 else "odd",L))
```

**Important Points**
- Returns map object → convert using list()
- Applies function to each element
- Faster and cleaner than loops
- Works with multiple iterables

**Filter**

**filter() returns only those elements from an iterable for which a given function returns True.**

```python
# Syntax

filter(function, iterable)
```

```python
def is_even(x):
    return x % 2 == 0

nums = [1, 2, 3, 4, 5]

result = filter(is_even, nums)

print(list(result))
```

```python
nums = [1, 2, 3, 4, 5]

result = list(filter(lambda x: x % 2 == 0, nums))

print(result)
```

```python
# numbers greater then 5

L = [3,4,5,6,7]
list(filter(lambda x: x>5,L))
```

```python
# fetch fruits starting with 'a'

fruits = ['apple','guava','cherry']

list(filter(lambda x: x.startswith('a'),fruits))
```

**Important Points**
- Returns a filter object (iterator)
- Convert using list()
- Works based on True/False condition
- Removes unwanted elements

**Reduce**

**reduce() applies a function cumulatively to the elements of an iterable and returns a single result.**

```python
# Syntax

from functools import reduce

reduce(function, iterable)
```

```python
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x + y, nums)

print(result)
```

```python
from functools import reduce

nums = [1, 2, 3, 4]

result = reduce(lambda x, y: x * y, nums)

print(result)
```

```python
from functools import reduce

nums = [1, 2, 3]

result = reduce(lambda x, y: x + y, nums, 10)

print(result)
```

```python
# sum of all item

import functools

functools.reduce(lambda x,y: x+y,[1,2,3,4,5])
```

```python
# find min

functools.reduce(lambda x,y: x if x<y else y,[23,11,45,10,1])
```

**Important Points**
- Must import from functools
- Returns a single value
- Works cumulatively
- Often used with lambda
