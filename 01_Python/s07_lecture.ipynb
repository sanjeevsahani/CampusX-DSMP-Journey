# **OOP**

**In Python, OOP (Object-Oriented Programming) is a programming paradigm that uses objects and classes to structure code and represent real-world entities.**

**Key Components**
- Class → blueprint
- Object → instance
- Attributes → data
- Methods → functions

**Why OOP is Used**
- Code becomes organized
- Easy to reuse
- Easy to maintain
- Models real-world problems

**Class - In Python, a class is a user-defined data structure that defines the properties (attributes) and behaviors (methods) of objects.**

```python
# Syntax

class ClassName:
    # attributes and methods
    pass
```

```python
class Student:
  pass

# creating object
s1 = Student()
print(s1)
```

```python
# Class with Attributes

class Student:
  def __init__(self, name, age):
    self.name = name
    self.age = age

s1 = Student("Sanjeev",22)

print(s1.name)
print(s1.age)
```

**Object - In Python, an object is an instance of a class that contains data (attributes) and behavior (methods).**

**Real-Life Example**

    Class = Car
    Object = BMW, Audi, Tesla

```python
class Student:
  pass

# Creating object
s1 = Student()
s2 = Student()

print(s1)
print(s2)
```

```python
# Object with Data

class Student:
  def __init__(self, name):
    self.name = name

s1 = Student("Aman")
s2 = Student("Sanjeev")

print(s1.name)
print(s2.name)
```

```python
# Object with Methods

class Student:
  def __init__(self, name):
    self.name = name

  def greet(self):
    print("Hello", self.name)

s1 = Student("Sanjeev")
s1.greet()
```

```python
# Counstructor + self

class Car:
  def __init__(self, brand, color):
    self.brand = brand
    self.color = color

  def show(self):
    print("Car:", self.brand, "-", self.color)

# Creating objects
c1 = Car("BMW", "Black")
c2 = Car("Audi", "White")

# Calling method
c1.show()
c2.show()
```

```python
class Atm:

  def __init__(self):
      self.pin = ''
      self.balance = 0
      self.menu()

  def menu(self):
      while True:
          user_input = input("""
          Hii how can I help you?
          1. Press 1 to create PIN
          2. Press 2 to change PIN
          3. Press 3 to check Balance
          4. Press 4 to withdraw
          5. Anything else to exit
          """)

          if user_input == '1':
              self.create_pin()
          elif user_input == '2':
              self.change_pin()
          elif user_input == '3':
              self.check_balance()
          elif user_input == '4':
              self.withdraw()
          else:
              print("Thank you!")
              break

  def create_pin(self):
      self.pin = input("Enter new PIN: ")
      self.balance = int(input("Enter balance: "))
      print("PIN created successfully")

  def change_pin(self):
      old_pin = input("Enter old PIN: ")
      if old_pin == self.pin:
          self.pin = input("Enter new PIN: ")
          print("PIN changed successfully")
      else:
          print("Wrong PIN")

  def check_balance(self):
      if input("Enter PIN: ") == self.pin:
          print("Balance:", self.balance)
      else:
          print("Wrong PIN")

  def withdraw(self):
      if input("Enter PIN: ") == self.pin:
          amount = int(input("Enter amount: "))
          if amount <= self.balance:
              self.balance -= amount
              print("Withdrawal successful. Balance:", self.balance)
          else:
              print("Insufficient balance")
      else:
          print("Wrong PIN")


# Run
atm = Atm()
```
