# **Topic - Output & Print Function**

**Output - In Python, we use the print() function to display output on the screen.**

```python
# python is a case sensitive language

print("Hello World")
```

**Text must always be inside single quotes ' ' or double quotes " ".**

```python
print("Sanjeev Sahani")
```

```python
print(Sanjeev Sahani)
```

```python
print(5)
```

**Numbers don't need quotes**

```python
print(3.14)
```

```python
print(True)
print(False)
```

```python
print("Hello",5,3.14,True)
```

**By default, values are separated by a space**

```python
print("Hello",5,3.14,True,sep="-")
```

**sep controls what goes between the values**

```python
print("Hello")
print("Sanjeev")
```

**By default, print() adds a new line at the end. Using end you can change this behavior.**

```python
print("Hello",end="-")
print("Sanjeev")
```

# **Topic - Variables**

**Variables - A variable is a name given to a memory location that stores a value in a program.**

```python
name = "sanjeev"
print(name)
```

```python
a = 5
b = 7

print(a + b)
```

**Static vs Dynamic Typing**

```python
# Dynamic Typing  --> Python dynamic typing suport karta hai

a = 5
```

```python
# static Typing  --> C / C++ / Java static typing suport karte hai

int a = 5
```

**Static vs Dynamic Binding**

```python
# Dynamic Binding

a = 5
print(a)
a = "Sanjeev"
print(a)
```

```python
# Static Binding

int a = 5;
printf("%d", a);   // print integer

char b = "Sanjeev";
printf("%s", b);   // print string
```

**Stylish Declaration Techniques**

```python
a = 1
b = 2
c = 3
print(a + b + c)
```

```python
a,b,c = 1,2,3
print(a,b,c)
```

```python
a = b = c = 5
print(a,b,c)
```
