**Arithmetic Operators - Operators used to perform mathematical calculations on numeric values.**

```python
# Addition
print(5 + 7)

# Subtraction
print(7 - 4)

# Multiplication
print(2 * 5)

# Division
print(5 / 2)

# Integer division (Floor division)
print(5 // 2)

# Modulus (Remainder)
print(9 % 2)

# Power (Exponent)
print(5 ** 2)
```

**Relational Operators - Operators used to compare two values and return a boolean result (True or False).**

```python
# Equal to
print(5 == 5)

# Not equal to
print(5 != 3)

# Greater than
print(7 > 4)

# Less than
print(3 < 8)

# Greater than or equal to
print(5 >= 5)

# Less than or equal to
print(4 <= 6)
```

**Logical Operators - Operators used to combine multiple conditions.**

```python
# AND Operator

print(1 and 0)

print(5 > 2 and 10 > 3)

print(5 > 10 and 8 > 3)
```

```python
# OR Operator

print(1 or 0)

print(5 > 10 or 8 > 3)

print(2 > 5 or 1 > 3)
```

```python
# NOT Operator

print(not 1)

print(not(5 > 3))

print(not(2 > 5))
```

```python
a = 10
b = 5

print(a > 5 and b < 10)
print(a < 5 or b < 10)
print(not(a == b))
```

**Bitwise Operators - Operators that perform operations on binary (bit-level) representations of integers.**

```python
# Bitwise AND &

print(5 & 3)
```

```python
# Bitwise OR |

print(5 | 3)
```

```python
# Bitwise XOR ^

print(5 ^ 3)
```

```python
# Bitwise NOT ~

print(~5)
```

```python
# Left Shift <<

print(5 << 1)
```

```python
# Right Shift >>

print(5 >> 1)
```

**Assignment Operators - Operators used to assign values to variables.**

```python
# assign value
x = 10

# add and assign
x += 5
print(x)

# subtract and assign
x -= 3
print(x)

# multiply and assign
x *= 2
print(x)

# divide and assign
x /= 2
print(x)

# modulus and assign
x %= 4
print(x)
```

**Membership Operators - Operators used to check whether a value exists in a sequence (list, string, etc.)**

```python
# in Operator

numbers = [1, 2, 3, 4, 5]

print(3 in numbers)
```

```python
text = "python"

print("p" in text)
```

```python
# not in Operator

numbers = [1, 2, 3, 4]

print(6 not in numbers)
```

```python
text = "python"

print("z" not in text)
```

**Program - Find the sum of a 3 digit number entered by the user.**

```python
# take the number from user
num = int(input("Enter a 3 digit number: "))

# extract digits
a = num % 10
num = num // 10

b = num % 10
num = num // 10

c = num % 10

# calculate the sum
total = a + b + c

# display the output
print("Sum =",total)
```

# **Topic - if-else**



*   if
*   else
*   elif





**if - The if statement is used to execute a block of code only when a specified condition is true.**

```python
# Basic Syntax

if condition:
    statement
```

```python
x = 10

if x > 5:
    print("x is greater than 5")
```

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Number is positive")
```

**else - The else statement executes a block of code when the if condition is false.**

```python
# Basic Syntax

if condition:
    statement1
else:
    statement2
```

```python
num = 5

if num > 0:
    print("Number is positive")
else:
    print("Number is negative")
```

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Number is even")
else:
    print("Number is odd")
```

**elif - The elif statement is used to check multiple conditions after the if statement.**

```python
# Basic Syntax

if condition1:
    statement1
elif condition2:
    statement2
else:
    statement3
```

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Number is positive")

elif num < 0:
    print("Number is negative")

else:
    print("Number is zero")
```

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")

elif marks >= 75:
    print("Grade B")

elif marks >= 50:
    print("Grade C")

else:
    print("Fail")
```

**Program - Find the minimum of 3 given numbers.**

```python
# take three numbers from user
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

# check which number is smallest
if a <= b and a <= c:
  minimum = a

elif b <= a and b <= c:
  minimum = b

else:
  minimum = c

# display the result
print("Minimum number =",minimum)
```

**Program - Menu Driven Calculator**

```python
# take numbers from user
num1 = int(input("Enter the 1st number: "))
num2 = int(input("Enter the 2nd number: "))

# take operation
op = input("Enter the operation: ")

# check operation
if op == "+":
  print(num1 + num2)

elif op == "-":
  print(num1 - num2)

elif op == "*":
  print(num1 * num2)

else:
  print(num1 / num2)
```

**Program - Menu Driven ATM.**

```python
menu = input("""
Hi! how can I help you.
1. Enter 1 for pin change
2. Enter 2 for balance check
3. Enter 3 for withdrawal
4. Enter 4 for exit
""")

if menu == "1":
  print("pin change")

elif menu == "2":
  print("balance check")

elif menu == "3":
  print("withdrawal")

elif menu == "4":
  print("exit")

else:
  print("Invalid option")
```
