**Q 1 - Write a Python function that takes a list and returns a new list with unique elements of the first list.**

    Input: [1,2,3,3,3,3,4,5]

    Output: [1, 2, 3, 4, 5]

```python
# function to return unique elements
def unique_list(L):
  result = []

  for i in L:
    # add omly if not already present
    if i not in result:
      result.append(i)


  return result

# input
L = [1,2,3,3,3,3,4,5]

print(unique_list(L))
```

**Q 2 - Write a Python function that accepts a hyphen-separated sequence of words as parameter and returns the words in a hyphen-separated sequence after sorting them alphabetically.**

    Input: green-red-yellow-black-white

    Output: black-green-red-white-yellow

```python
def sort_sequence(seq):
  temp = []

  # Split string and sort words
  for i in sorted(seq.split("-")):
    temp.append(i)

  # join back using hyphen
  return "-".join(temp)

# Input
s = "green-red-yellow-black-white"
sort_sequence(s)
```

**Q 3 - Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters**.


    Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'

    Expected Output :
        No. of Upper case characters :  9
        No. of Lower case Characters :  47

```python
# Function to count uppercase and lowercase
def count_case(s):
  upper = 0
  lower = 0

  # traverse each character
  for i in s:
    if i.isupper():
      upper += 1
    elif i.islower():
      lower += 1
    else:
      pass

  return lower, upper

# Input
s = "CampusX is an Online Mentorship Program fOr EnginEering studentS."

x, y = count_case(s)

print("No. of Lower case characters: ",x)
print("No. of Upper case characters: ",y)
```

**Q 4 - Write a Python program to print the even numbers from a given list.**

    Sample List : [1, 2, 3, 4, 5, 6, 7, 8, 9]
    
    Expected Result : [2, 4, 6, 8]

```python
# Function to return even numbers from a list
def even_list(L):
  result = []

  # Traverse list
  for i in L:
    if i%2 == 0:     # check even number
      result.append(i)

  return result

# Input
L = [1, 2, 3, 4, 5, 6, 7, 8, 9]

print(even_list(L))
```

**Q 5 - Write a Python function to check whether a number is perfect or not.**

    A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

    Example :
        The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
        Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

    The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.

```python
# Function to check perfect number
def is_perfect(n):
  sum_div = 0

  # Find divisors
  for i in range(1,n):
    if n % i == 0:
      sum_div += i

  # Check condition
  if sum_div == n:
    return "Perfect Number"
  else:
    return "Not Perfect Number"

# test
print(is_perfect(6))
print(is_perfect(28))
print(is_perfect(10))
```

**Q 6 - Write a Python function to concatenate any no of dictionaries to create a new one.**


    Sample Dictionary :
          dict1={1:10, 2:20}
          dict2={3:30, 4:40}
          dict3={5:50,6:60}

    Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

```python
# Function to merge any number of dictionaries
def merge_dicts(*kwargs):
  result = {}

  # Traverse each dictionary
  for i in kwargs:
    result.update(i)   # add key - value pairs

  return result

# Input
dict1 = {1:10, 2:20}
dict2 = {3:30, 4:40}
dict3 = {5:50, 6:60}

print(merge_dicts(dict1, dict2, dict3))
```

**Q 7 - Write a python function that accepts a string as input and returns the word with most occurence.**

    Input: hello how are you i am fine thank you

    Output: you -> 2

```python
def most_used(s):
  d = {}

  for i in s.split():
    if i in d:
      d[i] = d[i] + 1
    else:
      d[i] = 1

  # find maximum frequency
  max_val = (max(d.values()))

  # find word with max frequency
  for i in d:
    if d[i] == max_val:
      print(i,"->",d[i])
      break

most_used("hello how are you i am fine thank you")
```

**Q 8 - Write a python function that receives a list of integers and prints out a histogram of bin size 10.**

    Input: [13,42,15,37,22,39,41,50]

    Output: {11-20:2,21-30:1,31-40:2,41-50:3}

```python
# function to create histogram
def histogram(L):
  d = {}

  for i in L:
    start = (i // 10) * 10 + 1
    end = start + 9

    key = f"{start} - {end}"

    # count frequency
    if key in d:
      d[key] += 1
    else:
      d[key] = 1

  print(d)

# Input
L = [13,42,15,37,22,39,41,50]

histogram(L)
```

**Q 9 - Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.**

    List of Coordinates: [(1,1),(2,2),(3,3),(4,4)]
    Query Point: (0,0)

    Output: Nearest to (0,0) is (1,1)

```python
def shortest_dist(points,query):
  temp = []

  for i in points:
    distance = ((i[0] - query[0])**2 + (i[1] - query[1])**2)**0.5
    temp.append(distance)

  return points[sorted(list(enumerate(temp)),key=lambda x: x[1])[0][0]]

points = [(1,1),(2,2),(3,3),(4,4)]
query = (0,0)

shortest_dist(points,query)
```

**Q 10 - Write a python program that receives a list of strings and performs bag of word operation on those strings.**

    L = ["cat mat rat cat",
        "sat bat fat cat rat",
        "pat cat mat rat"]


```python
# Function for Bag of words
def bag_of_words(L):
  vocab = set()

  # create vocabulary
  for sentence in L:
    for word in sentence.split():
      vocab.add(word)

  vocab = sorted(vocab)

  result = []

  # count words for each sentence
  for sentence in L:
    word_count = {}
    words = sentence.split()

    for word in vocab:
      word_count[word] = words.count(word)

    result.append(word_count)

  return result

# input
L = ["cat mat rat cat",
    "sat bat fat cat rat",
    "pat cat mat rat"]

print(bag_of_words(L))
```

**Q 11 - Write a Python program to add three given lists using Python map and lambda.**

    Input: L1 = [1, 2, 3]
           L2 = [4, 5, 6]
           L3 = [7, 8, 9]

    Output: [12, 15, 18]

```python
# Input lists
L1 = [1, 2, 3]
L2 = [4, 5, 6]
L3 = [7, 8, 9]

# Using map and lambda
result = list(map(lambda x, y, z: x + y + z, L1, L2, L3))

print(result)
```

**Q 14 - Use reduce to convert a 2D list to 1D.**

    list1 = [[1, 2, 3],
             [3, 6, 7],
             [7, 5, 4]]

```python
from functools import reduce

# Input
list1 = [[1, 2, 3],
         [3, 6, 7],
         [7, 5, 4]]

# Using reduce to flatten
result = reduce(lambda x, y: x + y, list1)

print(result)
```

**Q 15 - A dictionary contains following information about 5 employees:**

- First name
- Last name
- Age
- Grade(Skilled,Semi-skilled,Highly skilled)

    Write a program using map/filter/reduce to a list of employees(first name + last name) who are highly skilled

```python
employees = [
    {"fname": "John", "lname": "Doe", "age": 30, "grade": "Skilled"},
    {"fname": "Jane", "lname": "Smith", "age": 28, "grade": "Highly skilled"},
    {"fname": "Mike", "lname": "Brown", "age": 35, "grade": "Semi-skilled"},
    {"fname": "Emma", "lname": "Davis", "age": 40, "grade": "Highly skilled"},
    {"fname": "Chris", "lname": "Wilson", "age": 25, "grade": "Skilled"}
]
```

```python
# Step 1: Filter highly skilled employees
filtered = filter(lambda emp: emp["grade"] == "Highly skilled", employees)

# Step 2: Map to full names
result = list(map(lambda emp: emp["fname"] + " " + emp["lname"], filtered))

print(result)
```
