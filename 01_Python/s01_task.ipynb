**Q 2 - Write a program that will convert celsius value to fahrenheit.**

```python
# take input from user
celsius = float(input("Enter the celsius value: "))

# use formula and calculate the value
fahrenheit = (9/5) * celsius + 32

# print the value
print("Fahrenheit =",fahrenheit,"F")
```

**Q 3 - Take 2 numbers as input from the user. Write a program to swap the numebrs without using any special python syntax.**

```python
# take input from user
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# swapping using temparory variable
temp = num1
num1 = num2
num2 = temp

# print the result
print("After swapping: ")
print("First number: ",num1)
print("Second number: ",num2)
```

**Q 4 - Write a program to find the euclidean distance between two coordinates. Take both the coordinates from the user as input.**

```python
# take input from user
x1 = int(input("Enter the value of x1: "))
y1 = int(input("Enter the value of y1: "))
x2 = int(input("Enter the value of x2: "))
y2 = int(input("Enter the value of y2: "))

# calculate euclidean distance
distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5

# display the result
print("Euclidean distance =",distance)
```

**Q 5 - Write a program to find the simple interest when the value of principal, rate of interest and time period is provided by the user.**

```python
# take principal amount from user
p = int(input("Enter the principal amount: "))

# take the rate if interest from user
r = int(input("Enter the rate of interest: "))

# takw the time period from user
t = int(input("Enter the time period (years): "))

# calculate the simple interest using formula
si = (p * r * t) / 100

# display the result
print("Simple interest is: ",si)
```

**Q 6 - Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.**

```python
# take input from user
heads = int(input("Enter total number of heads: "))
legs = int(input("Enter total number of legs: "))

# calculate number of dogs
dogs = (legs - 2 * heads) // 2

# calculate number of chichens
chickens = heads - dogs

# display the result
print("Number of Dogs: ",dogs)
print("Number of Chickens: ",chickens)
```

**Q 7 - Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.**

```python
# take input from user
n = int(input("Enter the value of n: "))

# calculate sum of squares
sum_squares = (n * (n + 1) * (2 * n + 1)) / 6

# display the result
print("Sum of squares =",sum_squares)
```

**Q 8 - Given the first 2 terms of an Arithmetic Series. Find the Nth term of the series. Assume all inputs are provided by the user.**

```python
# take the first and second term of the arithmetic series
a = int(input("Enter the 1st number: "))
b = int(input("Enter teh 2nd number: "))

# take which term we want to find
n = int(input("Enter the value of n: "))

# calculate common difference
d = b - a

# calculate the nth term using AP formula
an = a + (n - 1) * d

# display the result
print("The",n,"th term of the series is: ",an)
```

**Q 9 - Given 2 fractions, find the sum of those 2 fractions. Take the numerator and denominator values of the fractions from the user.**

```python
# take numerator and denominator from user
n1 = int(input("Enter the 1st numerator: "))
d1 = int(input("Enter the 1st denominator: "))
n2 = int(input("Enter the 2nd numerator: "))
d2 = int(input("Enter the 2nd denominator: "))

# calculate numerator
numerator = (n1 * d2) + (n2 * d1)

# calculate denominator
denominator = (d1 * d2)

# display the result
print("Sum =",numerator,"/",denominator)
```

**Q 10 - Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.**

```python
# take dimensions of the milk tank from the user
height = float(input("Enter the height of the tank: "))
width = float(input("Enter the width of the tank: "))
breadth = float(input("Enter the breadth of the tank: "))

# take volume of one glass
glass_volume = float(input("Enter the volume of one glass: "))

# calculate volume of the tank
tank_volume = height * width * breadth

# calculate number of glasses
glasses = tank_volume / glass_volume

# display result
print("Number of glasses that can be filled =", int(glasses))
```
