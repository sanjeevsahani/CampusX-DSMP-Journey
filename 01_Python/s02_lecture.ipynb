# **Topic - Operators**

**Arithmetic Operators - Operators used to perform mathematical calculations on numeric values.**

```python
# Addition
print(5 + 7)

# Subtraction
print(7 - 4)

# Multiplication
print(2 * 5)

# Division
print(5 / 2)

# Integer division (Floor division)
print(5 // 2)

# Modulus (Remainder)
print(9 % 2)

# Power (Exponent)
print(5 ** 2)
```

**Relational Operators - Operators used to compare two values and return a boolean result (True or False).**

```python
# Equal to
print(5 == 5)

# Not equal to
print(5 != 3)

# Greater than
print(7 > 4)

# Less than
print(3 < 8)

# Greater than or equal to
print(5 >= 5)

# Less than or equal to
print(4 <= 6)
```

**Logical Operators - Operators used to combine multiple conditions.**

```python
# AND Operator

print(1 and 0)

print(5 > 2 and 10 > 3)

print(5 > 10 and 8 > 3)
```

```python
# OR Operator

print(1 or 0)

print(5 > 10 or 8 > 3)

print(2 > 5 or 1 > 3)
```

```python
# NOT Operator

print(not 1)

print(not(5 > 3))

print(not(2 > 5))
```

```python
a = 10
b = 5

print(a > 5 and b < 10)
print(a < 5 or b < 10)
print(not(a == b))
```

**Bitwise Operators - Operators that perform operations on binary (bit-level) representations of integers.**

```python
# Bitwise AND &

print(5 & 3)
```

```python
# Bitwise OR |

print(5 | 3)
```

```python
# Bitwise XOR ^

print(5 ^ 3)
```

```python
# Bitwise NOT ~

print(~5)
```

```python
# Left Shift <<

print(5 << 1)
```

```python
# Right Shift >>

print(5 >> 1)
```

**Assignment Operators - Operators used to assign values to variables.**

```python
# assign value
x = 10

# add and assign
x += 5
print(x)

# subtract and assign
x -= 3
print(x)

# multiply and assign
x *= 2
print(x)

# divide and assign
x /= 2
print(x)

# modulus and assign
x %= 4
print(x)
```

**Membership Operators - Operators used to check whether a value exists in a sequence (list, string, etc.)**

```python
# in Operator

numbers = [1, 2, 3, 4, 5]

print(3 in numbers)
```

```python
text = "python"

print("p" in text)
```

```python
# not in Operator

numbers = [1, 2, 3, 4]

print(6 not in numbers)
```

```python
text = "python"

print("z" not in text)
```

# **Topic - if-else**



*   if
*   else
*   elif





**if - The if statement is used to execute a block of code only when a specified condition is true.**

```python
# Basic Syntax

if condition:
    statement
```

```python
x = 10

if x > 5:
    print("x is greater than 5")
```

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Number is positive")
```

**else - The else statement executes a block of code when the if condition is false.**

```python
# Basic Syntax

if condition:
    statement1
else:
    statement2
```

```python
num = 5

if num > 0:
    print("Number is positive")
else:
    print("Number is negative")
```

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Number is even")
else:
    print("Number is odd")
```

**elif - The elif statement is used to check multiple conditions after the if statement.**

```python
# Basic Syntax

if condition1:
    statement1
elif condition2:
    statement2
else:
    statement3
```

```python
num = int(input("Enter a number: "))

if num > 0:
    print("Number is positive")

elif num < 0:
    print("Number is negative")

else:
    print("Number is zero")
```

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")

elif marks >= 75:
    print("Grade B")

elif marks >= 50:
    print("Grade C")

else:
    print("Fail")
```

**Program - Find the minimum of 3 given numbers.**

```python
# take three numbers from user
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
c = int(input("Enter the third number: "))

# check which number is smallest
if a <= b and a <= c:
  minimum = a

elif b <= a and b <= c:
  minimum = b

else:
  minimum = c

# display the result
print("Minimum number =",minimum)
```

**Program - Menu Driven Calculator**

```python
# take numbers from user
num1 = int(input("Enter the 1st number: "))
num2 = int(input("Enter the 2nd number: "))

# take operation
op = input("Enter the operation: ")

# check operation
if op == "+":
  print(num1 + num2)

elif op == "-":
  print(num1 - num2)

elif op == "*":
  print(num1 * num2)

else:
  print(num1 / num2)
```

# **Topic - Loops**

*   While Loop
*   For Loop

**Loops - A loop is used to repeat a block of code multiple times until a condition becomes false. Loops help avoid writing the same code again and again.**

```python
print("Hello")
print("Hello")
print("Hello")
```

**While Loop - A while loop executes a block of code as long as the condition is true.**

```python
# Syntax

while condition:
    statement
```

```python
i = 1

while i <= 5:
    print(i)
    i = i + 1
```

```python
num = int(input("Enter the number: "))

i = 1

while i < 11:
  print(num,"x",i,"=",num * i)
  i += 1
```

**While Loop with if-else**

```python
# Syntax

while condition:
    if condition:
        statement
    else:
        statement
```

```python
i = 1

while i <= 5:
    if i % 2 == 0:
        print(i, "Even")
    else:
        print(i, "Odd")

    i = i + 1
```

```python
num = -3

while num <= 3:
    if num >= 0:
        print(num, "Positive")
    else:
        print(num, "Negative")

    num = num + 1
```

**Guessing Game Using While Loop**

```python
import random

# generate random number
jackpot = random.randint(1,100)

# take first guess
guess = int(input("Guess the number: "))
counter = 1

# run until correct guess
while guess != jackpot:
    if guess < jackpot:
        print("Wrong! Guess higher.")
    else:
        print("Wrong! Guess lower.")

    guess = int(input("Guess again: "))
    counter += 1

# correct guess message
print("Correct guess!")
print("Attempts:", counter)
```

**For Loop - A for loop is used to repeat a block of code a fixed number of times. It is commonly used with range(), lists, or strings.**

```python
# Syntax

for variable in range(start, stop, step):
    statement
```

```python
for i in range(1,6):
    print(i)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

**For Loop with if-else**

```python
# Syntax

for variable in sequence:
    if condition:
        statement
    else:
        statement
```

```python
for i in range(1, 6):
    if i % 2 == 0:
        print(i, "Even")
    else:
        print(i, "Odd")
```

```python
num = 5

for i in range(1, 11):
    result = num * i

    if result % 2 == 0:
        print(result, "Even")
    else:
        print(result, "Odd")
```

**Guessing Game Using For Loop**

```python
import random   # import random module

# generate random number between 1 and 100
jackpot = random.randint(1,100)

counter = 1   # variable to count attempts

# loop runs multiple times until break happens
for i in range(1,101):

    # take guess from user
    guess = int(input("guess karo: "))

    # check if guess is correct
    if guess == jackpot:
        print("correct guess")
        print("attempts:", counter)
        break   # stop the loop when correct

    # if guess is smaller
    elif guess < jackpot:
        print("galat! guess higher")

    # if guess is greater
    else:
        print("galat! guess lower")

    counter += 1   # increase attempt count
```
