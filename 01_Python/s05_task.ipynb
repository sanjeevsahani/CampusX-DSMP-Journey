# **Tuple**

**Q 1 - Join Tuples if similar initial element.**

    While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element.
    This problem can have applications in data domains such as Data Science.

    Example -
    
    Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

    Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
  

```python
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

unique = []

for i in test_list:
  unique.append(i[0])
unique = set(unique)

result = []
for i in unique:
  result.append([i])
  for j in test_list:
    if j[0] == i:
      result[-1].append(j[1])

print(list(map(tuple,result)))
```

**Q 4 - Count no of tuples, list and set from a list**
    
    list = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

    Output:
            List-2
            Set-2
            Tuples-1

# **Set**

**Q 1 - Write a program to find set of common elements in three lists using sets.**

    Input : ar1 = [1, 5, 10, 20, 40, 80]
            ar2 = [6, 7, 20, 80, 100]
            ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

    Output : [80, 20]

```python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

# Convert lists to sets
s1 = set(ar1)
s2 = set(ar2)
s3 = set(ar3)

# Find common elements using intersection
common = s1 & s2 & s3

# Convert to list
result = list(common)

print(result)
```

# **Dictionary**

**Q 1 - Key with maximum unique values**

    Given a dictionary with values list, extract key whose value has most unique values.

    Input: test_dict = {
        "CampusX" : [5, 7, 9, 4, 0],
        "is" : [6, 7, 4, 3, 3],
        "Best" : [9, 9, 6, 5, 5]
        }

    Output: CampusX


```python
test_dict = {
    "CampusX": [5, 7, 9, 4, 0],
    "is": [6, 7, 4, 3, 3],
    "Best": [9, 9, 6, 5, 5]
}

max_key = None
max_unique = 0

for key, value in test_dict.items():
    unique_count = len(set(value))

    if unique_count > max_unique:
        max_unique = unique_count
        max_key = key

print(max_key)
```

**Q 2 - Replace words from Dictionary. Given String, replace it’s words from lookup dictionary.**

    Input: test_str = 'CampusX best for DS students.'
           repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}

    Output: CampusX is the best channel for Data-Science students.

```python
test_str = 'CampusX best for DS students.'
repl_dict = {"best": "is the best channel", "DS": "Data-Science"}

words = test_str.split()

result = []

for word in words:
    if word in repl_dict:
        result.append(repl_dict[word])
    else:
        result.append(word)

final = " ".join(result)

print(final)
```

**Q 3 - Convert List to List of dictionaries. Given list values and keys list, convert these values to key value pairs in form of list of dictionaries**.

    Input: test_list = ["DataScience", 3, "is", 8]
           key_list = ["name", "id"]

    Output: [{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]


```python
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]

result = []

for i in range(0, len(test_list), 2):
    d = {
        key_list[0]: test_list[i],
        key_list[1]: test_list[i+1]
    }

    result.append(d)

print(result)
```

**Q 4 - Convert a list of Tuples into Dictionary.**


    Input: [("akash", 10), ("gaurav", 12), ("anand", 14), ("suraj", 20), ("akhil", 25), ("ashish", 30)]

    Output: {'akash': [10], 'gaurav': [12], 'anand': [14], 'suraj': [20], 'akhil': [25], 'ashish': [30]}

```python
test_list = [("akash", 10), ("gaurav", 12), ("anand", 14),
             ("suraj", 20), ("akhil", 25), ("ashish", 30)]

result = {}

for key, value in test_list:
    result[key] = [value]

print(result)
```

**Q 5 - Sort Dictionary key and values List.**

    Input: {'c': [3], 'b': [12, 10], 'a': [19, 4]}

    Output: {'a': [4, 19], 'b': [10, 12], 'c': [3]}

```python
D = {'c': [3], 'b': [12, 10], 'a': [19, 4]}

result = {}

for key in sorted(D.keys()):
    result[key] = sorted(D[key])

print(result)
```
