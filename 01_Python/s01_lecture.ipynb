```python
print("Hello",5,3.14,True,sep="-")
```

**Static vs Dynamic Binding**

```python
# Dynamic Binding

a = 5
print(a)
a = "Sanjeev"
print(a)
```

# **Topic - Type Conversion**

**Type Conversion - Type conversion is the process of converting a value from one data type to another.**

```python
# Implicit vs Explicit
```

**Implicit - Python automatically converts one data type to another.**

```python
a = 5
b = 5.7

t = a + b
print(t)

print(type(t))

print(type(a))
print(type(b))
```

**Explicit - Manually converting data types using functions.**

**Common Functions:**
- int() → convert to integer
- float() → convert to float
- str() → convert to string
- bool() → convert to boolean

```python
# str to int

x = "10"
y = int(x)
print(y)
print(type(y))
```

```python
# int to float

a = 5
b = float(a)
print(b)
print(type(b))
```

```python
# int to str

num = 100
text = str(num)
print(text)
print(type(text))
```

```python
# convert to bool

print(bool(0))   # False
print(bool(1))   # True
print(bool(""))  # False
```
