# **Tuples**

**A tuple is an ordered and immutable collection of elements. It is similar to a list, but the main difference is that tuple elements cannot be changed once they are created, whereas list elements can be modified.**

```python
T = (10, 20, 30, 20)

print(T[1])
```

**Characteristics of Tuples**

**1. Ordered**

- Elements are stored in a fixed sequence
- Each element has an index (position)
- ndex starts from 0

**2. Immutable (Unchangeable)**
- Once created, elements cannot be modified, added, or removed

```python
T = (1, 2, 3)
T[0] = 10   # Error
```

**3. Allows Duplicates**
- Same value can appear multiple times

```python
T = (1, 2, 2, 3)
```

**4. Heterogeneous**
- Can store different data types

```python
T = (1, "Hello", 3.5, True)
```

**Why Use Tuples - Tuples are used for storing fixed and unchangeable data efficiently, providing better performance and data safety.**

# **Topic - Difference between Lists and Tuples**

| Feature              | List                                | Tuple                         |
| -------------------- | ----------------------------------- | ----------------------------- |
| **Syntax**           | `[ ]`                               | `( )`                         |
| **Mutability**       | Mutable (can be changed)            | Immutable (cannot be changed) |
| **Speed**            | Slower                              | Faster                        |
| **Memory Usage**     | More memory                         | Less memory                   |
| **Built-in Methods** | Many (`append()`, `remove()`, etc.) | Few (`count()`, `index()`)    |
| **Error Prone**      | More (can change accidentally)      | Less (data is fixed)          |
| **Usability**        | Dynamic data                        | Fixed data                    |


**Use List when:**
- Data changes frequently
- Need to add/remove elements

**Use Tuple when:**
- Data should remain constant
- Need better performance
- Using as dictionary key

**Similarities**
- Both are ordered collections
- Allow duplicates
- Support indexing and slicing

```python
# List --> Mutability

a = [1, 2, 3]
b = a

a.append(4)
print(a)
print(b)
```

```python
# Tuple  --> Mutability

a = (1, 2, 3)
b = a

a = a + (4,)
print(a)
print(b)
```

```python
# speed

import time

L = list(range(1000000))
T = tuple(range(1000000))

start = time.time()
for i in L:
  i * 5
print("List time :",time.time()-start)

start = time.time()
for i in T:
  i * 5
print("Tuple time :",time.time()-start)
```

```python
# memory

import sys

L = list(range(1000))
T = tuple(range(1000))

print("List size: ",sys.getsizeof(L))
print("Tuple size: ",sys.getsizeof(T))
```

# **Topic - Accessing Elements in Tuples**

**In Python, a tuple elements are accessed using indexing and slicing, just like lists.**

**Positive Indexing**
- Each element has a position (index)
- Index starts from 0

```python
t = (10, 20, 30, 40, 50)

print(t[0])
print(t[1])
print(t[4])
```

**Negative Indexing**
- Used toaccess elements from the end

```python
t = (1, 2, 3, 4, 5)

print(t[-1])
print(t[-3])
print(t[-5])
```

**Slicing**
- Used to access a range of elements

```python
# positive

t = (1, 2, 3, 4, 5)

print(t[0:5])
print(t[1:3])
print(t[:])
print(t[:4])
print(t[2:])
print(t[::2])
```

```python
# Negative

t = (1, 2, 3, 4, 5)

print(t[-5:-1])
print(t[:-1])
print(t[-5:])
print(t[::-2])
```

```python
# inverse

t = (1, 2, 3, 4, 5)

print(t[::-1])
```

```python
# Nested tuple

t = (1, 2, (3, 4, 5))

print(t[1])
print(t[2])
print(t[2][0])
print(t[2][1])
print(t[2][2])
```

**Important Points**
- Use [] to access elements
- Index must be integer
- Out-of-range index → IndexError
- Tuples are immutable, but accessing is allowed

# **Topic - Adding Elements in Tuples**

**In Python, tuples are immutable, so you cannot directly add elements.**

```python
t = (1, 2, 3)
t.append(4)  # Error
```

**Tuples do not have methods like append() or insert()**

```python
# Adding itens using Concatenation (+)

t = (1, 2, 3)

t = t + (4,)
print(t)
```

```python
t = (1, 2, 3)

t = t + (4, 5, 6)
print(t)
```

**Important Points**
- Tuples are immutable → cannot modify directly

# **Topic - Deleting Elements in Tuples**

**In Python, Individual elements of a tuple cannot be deleted due to immutability, but the entire tuple can be deleted using the del statement.**

```python
t = (1, 2, 3)

del t[1]
```

**This deletes the whole tuple from memory**

```python
# Deleting Entire Tuple (Allowed)

t = (1, 2, 3)

del t
```

**Important Points**
- Cannot delete individual elements
- Can delete entire tuple using del
- Use list conversion for modifications

# **Topic - Operations on Tuples**

**In Python, tuples support operations like concatenation, repetiton, membership testing and iteration.**

**1. Concatenation (+)**
- Combines two tuples

```python
t1 = (1, 2)
t2 = (3, 4)

print(t1 + t2)
```

```python
t1 = (10, 20)
t2 = (30, 40, 50)

print(t1 + t2)
```

**2. Repetition (*)**
- Repeats elements of a tuple

```python
t = (1, 2, 3)

print(t * 2)
```

```python
t = (10, 20, 30)

print(t * 3)
```

**3. Membership (in, not in)**
- Checks if elements exists

```python
t = (10, 20, 30, 40, 50)

print(10 in t)
print(30 in t)

print(60 in t)
```

```python
t = (10, 20, 30, 40, 50)

print(10 not in t)
print(50 not in t)

print(70 not in t)
```

4. Iteration (Looping)
- Traverse elements

```python
t = (1, 2, 3, 4, 5)

for i in t:
  print(i)
```

**Important Points**
- Tuples support basic operations like lists
- But cannot be modified (immutable)
- Operations create new tuples

# **Topic - Tuple Functions**

**In Python, tuple functions like len(), max(), min(), and sum() perform operations on elements, while methods like count() and index() provide element-epecific information.**

```python
# len() - Returns number of elements

t = (1, 2, 3, 4, 5)

print(len(t))
```

```python
# max() - Returns the largest element

t = (10, 20, 30, 40, 50)

print(max(t))
```

```python
# min() - Returns the smallest element

t = (10, 5, 30, 2)

print(min(t))
```

```python
# sum() - Returns the sum of elements

t = (1, 2, 3, 4)

print(sum(t))
```

```python
# count() - returns number of occurrences of an element

t = (1, 2, 1, 3, 4, 1, 2)

print(t.count(1))
print(t.count(2))
print(t.count(3))
print(t.count(4))
```

```python
# index() - Returns index of the occurrence

t = (10, 20, 30)

print(t.index(10))
print(t.index(20))
```

```python
# sorted

t = (1, 2, 5, 3, 4)

print(sorted(t))
print(sorted(t,reverse=True))
```

```python
T = (5)    # Not a tuple, it's an integer
```

# **Sets**

**A set is an unordered collection of unique elements that supports mathematical operations like union, intersection, and difference.**

```python
S = {1, 2, 2, 3}

print(S)
```

**Characteristics of Sets**

**1. Unordered**
- No fixed order of elements
- Cannot access using index

**2. Unique Elements**
- Duplicate values are not allowed

**3. Mutable**
- Can add or remove elements
- But elements themselves must be immutable

**4. No Indexing or Slicing**

```python
S = {1, 2, 3}
# print(S[0]) Error
```

**5. Heterogeneous**

```python
S = {1, "Hello", 3.5}

print(S)
```

# **Topic - Creating Sets**

**A set is created using curly braces or the set() function, and it stores unique elements only.**

```python
# Empty set

S = set()   # ✔ Correct
print(S)
```

```python
S = {}      # This creates a dictionary, not a set
```

```python
# 1D Set

S = {1, 2, 3, 4}
print(S)
print(type(S))
```

```python
# 2D Set (Not allowed)

S = {1, 2, [3, 4]}   # Error
```

**Sets cannot contain mutable items like lists**

```python
S = {1, 2, (3, 4)}   # Allowed
print(S)
print(type(S))
```

```python
# Homogeneous Set --> All elements of same type

S = {1, 2, 3, 4, 5}
print(S)
```

```python
# Hetrogeneous Set --> Different data types allowed

S = {1, "Hello", 3.5}
print(S)
```

```python
# Using Type Conversion

S = set([1, 2, 3])        # list → set
print(S)
print(type(S))
```

```python
S = set((4, 5, 6))        # tuple → set
print(S)
print(type(S))
```

```python
S = set("Hello")          # string → set
print(S)
print(type(S))
```

```python
# Duplicates Not Allowed

S = {1, 2, 2, 3}

print(S)
```

**Duplicate values are automatically removed**

**Important Points**
- Use set() for empty set
- No duplicates allowed
- Only immutable elements allowed
- Unordered collection

# **Topic - Accessing Elements in Sets**

**In Python, a sets are unordered collections, so elements cannot be accessed using indexing; instead, they are accessed using loops or membership operations.**

```python
# Indexing (Not allowed)

S = {10, 20, 30}

print(S[0])   # Error
```

```python
# Accessing Using Loop (Correct Way)

S = {10, 20, 30}

for item in S:
    print(item)
```

```python
# Membership Testing

S = {10, 20, 30}

print(20 in S)      # True
print(50 not in S)  # True
```

```python
# Converting Set to List (For Indexing)

S = {10, 20, 30}

L = list(S)
print(L[0])   # Now indexing works
```

**Important Points**
- No indexing or slicing
- Use loop for traversal
- Use in for checking elements
- Order is not guaranteed

# **Topic - Editing Elements in Sets**

**Changing Elements (Not Directly Possible)**
- Sets are unordered
- No indexing → cannot update directly

```python
S = {1, 2, 3}

S[0] = 10   # Error
```

# **Topic - Adding Elements in Sets**

**In Python, Elements can be added to a set using add() for single items and update() for multiple items, while ensuring elements are unique and immutable.**

**add() - Adds a single element to the set**

```python
S = {1, 2, 3}

S.add(4)
print(S)
```

**update() - Adds multiple elements at once. Accepts iterable (list, tuple, set, etc.)**

```python
S = {1, 2}

S.update([3, 4, 5])
print(S)
```

```python
# using multiple iterables

S = {1, 2}

S.update([3, 4], (5, 6))
print(S)
```

**Adding Duplicate Values**

```python
S = {1, 2, 3}

S.add(2)
print(S)
```

**Adding Immutable Elements Only**

```python
S = {1, 2}

S.add((3, 4))   # Allowed (tuple)
print(S)
```

```python
S = {1, 2}

S.add([5, 6])   # Error (list is mutable)
```

**Important Points**
- Use add() → for single element
- Use update() → for multiple elements
- Duplicates are automatically ignored
- Only immutable elements can be added

# **Topic - Deleting Elements in Sets**

**In Python, Elements in a set can be removed using remove(), discard(), pop(), and clear(), depending on the requirement.**

**remove() - Removes a specific element**

```python
S = {1, 2, 3}

S.remove(2)
print(S)
```

```python
# Error if element not found:

S = {1, 2, 3}

S.remove(5)   # KeyError
print(S)
```

**discard() -**
- Removes element if present
- No error if element not found

```python
S = {1, 2, 3}

S.discard(2)
print(S)

S.discard(5)
print(S)
```

**pop() - Removes a random element**

```python
S = {10, 20, 30}

print(S.pop())
```

**clear() - Removes all elements**

```python
S = {1, 2, 3}

S.clear()
print(S)
```

**del() - Deletes the entire set**

```python
S = {1, 2, 3}

del S
```

**Important Differences**
| Method    | Removes       | Error if not found |
| --------- | ------------- | ------------------ |
| remove()  | Specific item | Yes                |
| discard() | Specific item | No                 |
| pop()     | Random item   | No                 |
| clear()   | All items     | No                 |


**Important Points**
- Sets are unordered, so pop() removes random element
- Use discard() for safe deletion
- Use clear() to empty set
