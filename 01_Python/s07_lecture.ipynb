# **OOP**

**In Python, OOP (Object-Oriented Programming) is a programming paradigm that uses objects and classes to structure code and represent real-world entities.**

**Key Components**
- Class → blueprint
- Object → instance
- Attributes → data
- Methods → functions

**Why OOP is Used**
- Code becomes organized
- Easy to reuse
- Easy to maintain
- Models real-world problems

```python
class Atm:

  def __init__(self):
      self.pin = ''
      self.balance = 0
      self.menu()

  def menu(self):
      while True:
          user_input = input("""
          Hii how can I help you?
          1. Press 1 to create PIN
          2. Press 2 to change PIN
          3. Press 3 to check Balance
          4. Press 4 to withdraw
          5. Anything else to exit
          """)

          if user_input == '1':
              self.create_pin()
          elif user_input == '2':
              self.change_pin()
          elif user_input == '3':
              self.check_balance()
          elif user_input == '4':
              self.withdraw()
          else:
              print("Thank you!")
              break

  def create_pin(self):
      self.pin = input("Enter new PIN: ")
      self.balance = int(input("Enter balance: "))
      print("PIN created successfully")

  def change_pin(self):
      old_pin = input("Enter old PIN: ")
      if old_pin == self.pin:
          self.pin = input("Enter new PIN: ")
          print("PIN changed successfully")
      else:
          print("Wrong PIN")

  def check_balance(self):
      if input("Enter PIN: ") == self.pin:
          print("Balance:", self.balance)
      else:
          print("Wrong PIN")

  def withdraw(self):
      if input("Enter PIN: ") == self.pin:
          amount = int(input("Enter amount: "))
          if amount <= self.balance:
              self.balance -= amount
              print("Withdrawal successful. Balance:", self.balance)
          else:
              print("Insufficient balance")
      else:
          print("Wrong PIN")

# Run
atm = Atm()
```

# **Methods and Functions**

**Function - A function is a standalone block of code that preforms a specific task.**

```python
def add(a, b):
  return a + b

print(add(2, 3))
```

**Key Features**
- Defined outside any class
- Can be called directly
- Works on any data passed

**Method - A method is a function that is defined inside a class and is associated with an object.**

```python
class Calculator:
  def add(self, a, b):
    return a + b

c = Calculator()
print(c.add(2, 3))
```

**Key Features**
- Defined inside a class
- Called using an object
- Works with object data using self

**Differences**

| Feature         | Function        | Method                |
| --------------- | --------------- | --------------------- |
| Defined in      | Outside class   | Inside class          |
| Binding         | Independent     | Bound to object/class |
| First parameter | Normal          | `self` / `cls`        |
| Call            | `func()`        | `obj.method()`        |
| Usage           | General purpose | Object-specific       |


```python
L = [1, 2, 3]

len(L) # function -> Because it is outside the list class
L.append() # method -> Because it is inside the list class
```

    len() - built-in function
    append() - list method

**Important**

    Every method is a function
    But every function is not a method

# **Types of Methods in Python**

**In Python, methods inside a class are mainly of three types:**

**1. Instance Method - Works with object (instance) data.**

```python
class Student:
  def __init__(self, name):
    self.name = name

  def show(self):
    print("Name:", self.name)

s = Student("Sanjeev")
s.show()
```

**2. Class Method - Works with class-level data (shared by all objects)**

```python
class Student:
  college = "ABC College"

  @classmethod
  def show_college(cls):
    print("College:", cls.college)

Student.show_college()
```

**3. Static Method - Does not depend on class or object data**

```python
class Math:
  @staticmethod
  def add(a, b):
    return a + b

print(Math.add(2, 3))
```

**Comparison Table**

| Feature         | Instance Method | Class Method     | Static Method    |
| --------------- | --------------- | ---------------- | ---------------- |
| Works with      | Object data     | Class data       | No data          |
| First parameter | `self`          | `cls`            | None             |
| Decorator       | No              | `@classmethod`   | `@staticmethod`  |
| Call            | `obj.method()`  | `Class.method()` | `Class.method()` |


```python
class Demo:

    def instance_method(self):      # instance method
        pass

    @classmethod
    def class_method(cls):          # class method
        pass

    @staticmethod
    def static_method():            # static method
        pass
```

# **Magic Methods (Dunder Methods)**

**In Python, magic methods are special methods that start and end with double underscores (__) used to define how objects behave with built-in operations.**

    Double + Under → Dunder

        __init__
        __str__
        __len__
        __add__
        __eq__

```python
# 1) __init__ → Constructor

class A:
  def __init__(self):
    print("Object created")
```

```python
# 2) __str__ → String Representation

class Student:
  def __init__(self, name):
    self.name = name

  def __str__(self):
    return self.name

s = Student("Sanjeev")
print(s)
```

```python
# 3) __len__ → Length

class MyList:
  def __init__(self, data):
    self.data = data

  def __len__(self):
    return len(self.data)

obj = MyList([1, 2, 3])
print(len(obj))
```

```python
# 4) __add__ → + Operator

class Number:
  def __init__(self, x):
    self.x = x

  def __add__(self, other):
    return self.x + other.x

n1 = Number(10)
n2 = Number(20)

print(n1 + n2)
```

```python
# 5) __eq__ → == Operator

def __eq__(self, other):
  return self.x == other.x
```

**Important Points**
- Always start & end with __
- Automatically called by Python
- Not called directly usually
