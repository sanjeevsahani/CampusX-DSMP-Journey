# **Topic - Nested Loops**

**Nested Loop - A nested loop is simply a loop inside another loop. The inner loop runs completely every time the outer loop runs once.**

```python
# Example --> unique pairs

for i in range(1,4):   # outer loop
  for j in range(1,4):  #inner loop
    print(i,j)
```

**Pattern 1**

    *
    * *
    * * *

```python
# take number of rows from user
rows = int(input("Enter the rows: "))

# outer loop controls the number of rows
for i in range(1,rows+1):

  # inner loop prints stars in each row
  for j in range(1,i+1):
    print("*",end=" ")

  # move to next line
  print()
```

**Pattern 2**

    1
    1 2 1
    1 2 3 2 1
    1 2 3 4 3 2 1

```python
# take number of rows from user
rows = int(input("Enter the rows: "))

# outer loop controls the number of rows
for i in range(1,rows+1):

  # increasing numbers
  for j in range(1,i+1):
    print(j,end=" ")

  # decreasing numbers
  for k in range(i-1,0,-1):
    print(k,end=" ")

  # move to next line
  print()
```

# **Topic - Loop Control Statement**

- Break
- Continue
- Pass

**Break - Stops the loop immediately, even if the loop condition is still true.**

```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input("Enter lower range: "))
upper = int(input("Enter upper range: "))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break

  else:
    print(i)
```

**Continue - Skips the current iteration and moves to the next iteration of the loop.**

```python
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
# print only odd numbers

for i in range(1, 10):

    # check if number is even
    if i % 2 == 0:
        continue

    print(i)
```

**Pass - Does nothing. It is used as a placeholder when a statement is required but no action is needed.**

```python
for i in range(1,10):
  pass
```

# **Creating Strings in Python**

**Creating a string means storing text inside quotes in Python.**

```python
# Using Single Quotes

text = 'Hello'
print(text)
```

```python
# Using Double Quotes

text = "Hello"
print(text)

msg = "It's Python"
print(msg)
```

```python
# Using Triple Single Quotes

text = '''This is
a multi-line
string'''

print(text)
```

```python
# Using Triple Double Quotes

text = """This is
a multi-line
string"""

print(text)
```

```python
# Using str() Function

num = 5
text = str(num)

print(text)
```

```python
# Empty String

text = ""
```

```python
# String with Special Characters

text = "Hello\nSanjeev"
print(text)
```

```python
# Raw String

text = r"Hello\nSanjeev"
print(text)
```

# **Accessing Strings in Python**

**Accessing a string means retrieving characters or parts of a string using indexing and slicing.**

```python
# Postive Indexing (Access Single Character)

text = "Sanjeev"

print(text[0])
print(text[3])
```

```python
# Negative Indexing

text = "Sanjeev"

print(text[-1])
print(text[-7])
```

```python
# Slicing (Access Multiple Characters)

text = "Sanjeev"

print(text[0:4])
print(text[3:])
print(text[:4])
```

```python
# Negative Slicing

text = "Sanjeev"

print(text[-7:-1])
print(text[:-1])
print(text[-7:])
```

```python
# Slicing with Step

text = "Sanjeev"

print(text[0:7:2])
```

```python
# Reverse String

text = "Sanjeev"

print(text[::-1])
```

# **Adding Characters to Strings in Python**

**Python strings are immutable, you cannot change them directly. Instead, you create a new string by adding characters.**

```python
# Using + (Concatenation)

text_01 = "Hey"
text_02 = " Sanjeev"

text = text_01 + text_02
print(text)
```

```python
# Adding a Singl Character

text = "Hell"

text = text + "o"

print(text)
```

# **Deleting Strings in Python**

In Python, you can delete an entire string variable, but you cannot delete individual characters (because strings are immutable).

```python
# Deleting Entire String

text = "Python"
del text
print(text)  # Error
```

```python
# Deleting Part of a String (Not Directly Possible)

text = "Python"
del text[0]   # ERROR
```

```python
# Correct Way (Create New String)

text = "Python"

# remove first character
text = text[1:]

print(text)
```

# **Editing Strings in Python**

In Python, strings are immutable, which means you cannot change them directly.
To “edit” a string, you must create a new string.

```python
# Direct Editing (Not Allowed)

text = "Hello"
text[0] = "Y"   # Error
```

```python
# Editing Using Slicing

text = "Hello"

# change 'H' to 'Y'
text = "Y" + text[1:]

print(text)
```

```python
# Replacing Characters or Words

text = "Hello World"

text = text.replace("World", "Sanjeev")

print(text)
```

# **Operations on Strings**

- Arithmetic Operations
- Relational Operations
- Logical Operations
- Membership Operations
- Loops on Strings

**Arithmetic Operations**

Arithmetic operations on strings are + (concatenation) and * (repetition).

```python
# Concatenation (+) - Used to join two or more strings.

a = "Hello"
b = "Sanjeev"

print(a + " " + b)
```

```python
# Repetition (*) - Used to repeat a string multiple times.

print("Hi " * 3)
```

```python
print("*" * 50)
```

**Relational Operations**

Relational operations compare two strings based on their character values.

```python
print("delhi" == "delhi")

print("ballia" != "ballia")

print("apple" < "banana")

print("apple" > "banana")
```

**Logical Operations**

Logical operations combine multiple conditions involving strings using and, or, and not.

```python
# and Operator

"hello" and "world"
```

```python
# or Operator

"hello" or "world"
```

```python
# not Operator

not "hello"
```

**Membership Operations**

Membership operations check whether a value exists in a string using in and not in.

```python
# Using in

text = "Python"

print("Py" in text)
print("on" in text)
print("Java" in text)
```

```python
# Using not in

text = "Python"

print("Java" not in text)
print("Py" not in text)
```

**Loops on Strings**

Looping on strings means accessing each character one by one using loops.

```python
for i in "Sanjeev":
  print(i)
```

```python
for i in "pune":
  print("Ballia")
```
