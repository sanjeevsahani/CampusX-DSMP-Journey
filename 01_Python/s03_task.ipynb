**Q 1 - Print the following pattern. Writea program to use for loop to print the following reverse number pattern.**

    5 4 3 2 1
    4 3 2 1
    3 2 1
    2 1
    1

```python
# take number of rows from user
rows = int(input("Enter number of rows: "))

# outer loop controls the number of rows
for i in range(0,rows):

  # inner loop prints the numbers
  for j in range(rows-i,0,-1):
    print(j,end=" ")

  # move to next line
  print()
```

**Q 2 - Print the follwing pattern.**

    *
    * *
    * * *
    * * * *
    * * * * *
    * * * *
    * * *
    * *
    *

```python
# take number of rows
rows = int(input("Enter number of rows: "))

# upper part (increasing)
for i in range(1,rows+1):
  for j in range(1,i+1):
    print("*",end=" ")
  print()

# lower part (decreasing)
for i in range(1,rows):
  for j in range(rows-i,0,-1):
   print("*",end=" ")
  print()
```

**Q 5 - Write a Python Program to Find the Sum of the Series till the nth term:**

    1 + x^2/2 + x^3/3 + ..... x^n/n
    n will be provided by thw user

```python
# take input from user
x = int(input("Enter the value of x: "))
n = int(input("Enter the value of n: "))

total = 1       # first term of series
series = '1'    # to display series

# loop from 2 to n
for i in range(2,n+1):
  term = (x ** i) / i  # calculate each term
  total += term        # add to total

  # build series string
  series += " + {}^{}/{}".format(x, i, i)

# Print result
print("Series: ",series)
print("Sum: ",total)
```

**Q 6 - Find the sum of the series upto n terms.**

    Write a program to calculate the sum of series up to n term.
    For example, if n = 5 the series will become 2 + 22 + 222 + 2222 + 22222 = 24690.
    Take the user input and then calculate. And the output style should match which is given in the example.

    Example:-
    
      Input - 5

      Output - 2 + 22 + 222 + 2222 + 22222
      Sum of above series is : 24690

```python
# take input from user
n = int(input("Enter the value of n: "))

start = 2   # first term
total = 0   # to store total

# print series
for i in range(0,n):
  if i < n-1:
    print(start,end=" + ")
  else:
    print(start)

  total = total + start   # add term to sum
  start = start * 10 + 2  # generate next term

print("Sum of above series is: ",total)
```

**Q 7 - Write a program to print all the unique combinations of 1, 2, 3 and 4.**

    Output:- 1 2 3 4
             1 2 4 3

             1 4 2 3
             1 4 3 2
             2 1 3 4
             2 1 4 3
             2 3 1 4
             2 3 4 1
             1 4 1 3
             .
             .
             and so on

```python
# numbers
nums = [1, 2, 3, 4]

# generate all unique combinations
for i in nums:
    for j in nums:
        for k in nums:
            for l in nums:

                # ensure all are different
                if i != j and i != k and i != l and j != k and j != l and k != l:
                    print(i, j, k, l)
```

**Q 8 - Write a program that will take a decimal number as input and prints out the binary equivalent of the number**

```python
# take input from user
n = int(input("Enter the number: "))

binary = []   # list to store remainders

# convert to binary
while n > 0:
    binary.append(n % 2)   # store remainder
    n = n // 2             # update number

# print in reverse
for i in binary[::-1]:
    print(i, end="")
```

**Q 9 - Write a program that will take 2 numbers as input and prints the LCM and HCF of those 2 numbers.**

```python
a = int(input("Enter the 1st number: "))
b = int(input("Enter the 2nd number: "))

# find HCF
x, y = a, b
while y != 0:
    x, y = y, x % y

hcf = x

# find LCM
greater = max(a, b)

while True:
    if greater % a == 0 and greater % b == 0:
        lcm = greater
        break
    greater += 1

print("HCF:", hcf)
print("LCM:", lcm)
```

**Q 13 - Take a alphanumeric string input and print the sum and average of the digits that appear in the string, ignoring all other characters.**

    Input - hel1234every093

    Output - Sum: 22
             Avg: 3.1428

```python
# take input
text = input("Enter the string: ")

total = 0
count = 0

# loop through string
for i in text:

    if i.isdigit():          # check if character is digit
        total += int(i)      # convert to int and add
        count += 1            # count digits

# calculate average
if count > 0:
    avg = total / count
else:
    avg = 0

# print result
print("Sum:", total)
print("Avg:", avg)
```

**Q 15 - Check whether the string is Symmetrical.**

    Input - khokho

    Output - The entered string is symmetrical

```python
# take input
text = input("Enter the string: ")

# divide string into two halves
if len(text) % 2 == 0:
    first = text[:len(text)//2]
    second = text[len(text)//2:]
else:
    first = text[:len(text)//2]
    second = text[len(text)//2+1:]

# check symmetry
if first == second:
    print("The entered string is symmetrical")
else:
    print("The entered string is not symmetrical")
```

**Q 17 - Find uncommon words from two String.**

    Statement: Given two sentences as string A and B. The task is to return a list of all uncommon words.
    A word is uncommon if it appears exactly once in any one of the sentences, and dose not appear in the other sentence.
    Note: A sentence is a string of space-separated words. Each word consists only of lowercase letters.

    Input - A = "apple banana mango"
            B = "banana fruits mango"

    Output - ['apple','fruits']

```python
# given strings
A = input("Enter first string: ")
B = input("Enter second string: ")

# convert strings into word lists
listA = A.split()
listB = B.split()

result = []

# check words from first string
for word in listA:
    if listA.count(word) == 1 and word not in listB:
        result.append(word)

# check words from second string
for word in listB:
    if listB.count(word) == 1 and word not in listA:
        result.append(word)

# print result
print(result)
```

**Q 18 - Find a location of a word in a given String sentence. Don't use index / find functions.**

    Input -
       sentence: We can learn data science through campusx mentorship program.
       word: campusx

    Output - Location of the word is 7

```python
# take input
sentence = input("Enter sentence: ")
word = input("Enter word to find: ")

# split sentence into words
words = sentence.split()

# search manually
for i in range(len(words)):
    if words[i] == word:
        print("Location of the word is", i + 1)
        break
else:
    print("Word not found")
```

**Q 19 - Write a program that can remove all the duplicate characters from a string. User will provide the input.**

```python
# take input
text = input("Enter the string: ")

result = ""

# loop through characters
for ch in text:

    if ch not in result:   # check if already added
        result += ch       # add only once

print(result)
```
