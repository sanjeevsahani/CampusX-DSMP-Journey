# **Topic - Nested Loops**

**Nested Loop - A nested loop is simply a loop inside another loop. The inner loop runs completely every time the outer loop runs once.**

```python
# Example --> unique pairs

for i in range(1,4):   # outer loop
  for j in range(1,4):  #inner loop
    print(i,j)
```

**Pattern 1**

    *
    * *
    * * *

```python
# take number of rows from user
rows = int(input("Enter the rows: "))

# outer loop controls the number of rows
for i in range(1,rows+1):

  # inner loop prints stars in each row
  for j in range(1,i+1):
    print("*",end=" ")

  # move to next line
  print()
```

**Pattern 2**

    1
    1 2 1
    1 2 3 2 1
    1 2 3 4 3 2 1

```python
# take number of rows from user
rows = int(input("Enter the rows: "))

# outer loop controls the number of rows
for i in range(1,rows+1):

  # increasing numbers
  for j in range(1,i+1):
    print(j,end=" ")

  # decreasing numbers
  for k in range(i-1,0,-1):
    print(k,end=" ")

  # move to next line
  print()
```

# **Topic - Loop Control Statement**

- Break
- Continue
- Pass

**Break - Stops the loop immediately, even if the loop condition is still true.**

```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input("Enter lower range: "))
upper = int(input("Enter upper range: "))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break

  else:
    print(i)
```

**Continue - Skips the current iteration and moves to the next iteration of the loop.**

```python
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
# print only odd numbers

for i in range(1, 10):

    # check if number is even
    if i % 2 == 0:
        continue

    print(i)
```

**Pass - Does nothing. It is used as a placeholder when a statement is required but no action is needed.**

```python
for i in range(1,10):
  pass
```

# **Accessing Strings in Python**

**Accessing a string means retrieving characters or parts of a string using indexing and slicing.**

```python
# Postive Indexing (Access Single Character)

text = "Sanjeev"

print(text[0])
print(text[3])
```

```python
# Negative Indexing

text = "Sanjeev"

print(text[-1])
print(text[-7])
```

```python
# Slicing (Access Multiple Characters)

text = "Sanjeev"

print(text[0:4])
print(text[3:])
print(text[:4])
```

```python
# Negative Slicing

text = "Sanjeev"

print(text[-7:-1])
print(text[:-1])
print(text[-7:])
```

```python
# Slicing with Step

text = "Sanjeev"

print(text[0:7:2])
```

```python
# Reverse String

text = "Sanjeev"

print(text[::-1])
```

# **Deleting Strings in Python**

In Python, you can delete an entire string variable, but you cannot delete individual characters (because strings are immutable).

```python
# Deleting Entire String

text = "Python"
del text
print(text)  # Error
```

```python
# Deleting Part of a String (Not Directly Possible)

text = "Python"
del text[0]   # ERROR
```

```python
# Correct Way (Create New String)

text = "Python"

# remove first character
text = text[1:]

print(text)
```

# **Editing Strings in Python**

In Python, strings are immutable, which means you cannot change them directly.
To “edit” a string, you must create a new string.

```python
# Direct Editing (Not Allowed)

text = "Hello"
text[0] = "Y"   # Error
```

```python
# Editing Using Slicing

text = "Hello"

# change 'H' to 'Y'
text = "Y" + text[1:]

print(text)
```

```python
# Replacing Characters or Words

text = "Hello World"

text = text.replace("World", "Sanjeev")

print(text)
```

**Arithmetic Operations**

Arithmetic operations on strings are + (concatenation) and * (repetition).

```python
# Concatenation (+) - Used to join two or more strings.

a = "Hello"
b = "Sanjeev"

print(a + " " + b)
```

**Relational Operations**

Relational operations compare two strings based on their character values.

```python
print("delhi" == "delhi")

print("ballia" != "ballia")

print("apple" < "banana")

print("apple" > "banana")
```

**Logical Operations**

Logical operations combine multiple conditions involving strings using and, or, and not.

```python
# and Operator

"hello" and "world"
```

```python
# or Operator

"hello" or "world"
```

```python
# not Operator

not "hello"
```

**Membership Operations**

Membership operations check whether a value exists in a string using in and not in.

```python
# Using in

text = "Python"

print("Py" in text)
print("on" in text)
print("Java" in text)
```

```python
# Using not in

text = "Python"

print("Java" not in text)
print("Py" not in text)
```

# **Topic - Program Problem**

**Q 1 - Find the length of a given string without using the len() function.**

```python
# take string from user
text = input("Enter the string: ")

counter = 0

for i in text:
  counter += 1

# print the final length
print("Length of string is ",counter)
```

**Q 2 - Extract username from a given email. Eg if the email is sanjeev@gmail.com. Then the username should be sanjeev.**

```python
# take email from user
email = input("Enter email: ")

# find position of the '@'
pos = email.index("@")

# extract and print the username
print(email[0:pos])
```

**Q 3 - Count the frequency of a particular character in a provided string. Eg 'hello how are you' is the string, the frequency of h in string is 2.**

```python
# take string and character from user
text = input("Enter the strings: ")
term = input("Enter the character to count: ")

count = 0

for i in text:
  if i == term:
    count += 1

print("Frequency :",count)
```

**Q 4 - Write a program which can remove a particular character from a string.**

```python
# take string and character from user
text = input("Enter the strings: ")
term = input("Enter character to remove: ")

result = ""

# loop through each character
for i in text:
  if i != term:
    result += i

print(result)

```

**Q 5 - Write a program that can check whether a given string is palindrome or not. Eg - abba, malayalam.**

```python
# take input from user
text = input("Enter a string: ")

# check palindrome
if text == text[::-1]:
  print("Palindrome")
else:
  print("Not Palindrome")
```

**Q 6 - Write a program to count the numebr of words in a string without split().**

```python
text = input("Enter the string: ")
L = []
temp = ""
for i in text:

  if i != " ":
    temp = temp + i
  else:
    L.append(temp)
    temp = ""

L.append(temp)
print(L)
```

**Q 7 - Write a python program to convert a string to title case without using the title().**

```python
# take string from user
text = input("Enter the string: ")

L = []
for i in text.split():
  L.append(i[0].upper() + i[1:].lower())

print(" ".join(L))
```

**Q 8 - Write a program that can convert an integer to string, without using str().**

```python
number = int(input("Enter the number: "))

digits = "0123456789"
result = ""

# handle zero case
if number == 0:
    result = "0"
else:
    while number != 0:
        result = digits[number % 10] + result
        number = number // 10

print(result)
print(type(result))
```
